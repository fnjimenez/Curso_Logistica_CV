## ⚙️ **Paso 6: Solver de Optimización Logística** 🧮

### 🎯 **Objetivo del Solver**
Encontrar la **distribución óptima** de productos desde los HUBs hacia las ciudades destino, minimizando costos totales bajo restricciones operativas y presupuestales.

### 🔧 **Componentes del Modelo**
| Elemento | Tipo | Descripción |
|----------|------|-------------|
| **Variables Flujo** | Entero | Unidades a transportar por ruta |
| **Variables Activación** | Binario | Rutas activas (costo fijo) |
| **Función Objetivo** | Minimizar | Costo transporte + costos fijos |
| **Restricciones** | Lineales | Capacidad, demanda, presupuesto |

### 🚦 **Restricciones Implementadas**
- **🏭 Capacidad HUBs** - No exceder producción
- **🎯 Demanda Ciudades** - Satisfacer necesidades
- **💰 Presupuesto Máximo** - Límite financiero
- **🔄 Activación Rutas** - Relación flujo-binario
- **🎯 Restricciones Especiales** - Por módulo del alumno

---

**🚀 Solver configurado para optimización avanzada**

```python
# ============================================================================
# BLOQUE CODIGO 06  SOLVER DE OPTIMIZACIÓN (VERSIÓN COMPLETA Y CORREGIDA)
# ============================================================================
print("\n⚙️ CONFIGURANDO SOLVER OPTIMIZADO Y COMPLETO...")
print("=" * 50)

# Importar módulo time para medir duración
import time

def optimizar(oferta, demanda, costos, mod5, presupuesto_max, costo_fijo_trailer=COSTO_FIJO_TRAILER):
    """
    SOLVER COMPLETO Y CORREGIDO - Incluye costos fijos, variables binarias y validaciones
    """
    print("   🚀 Iniciando optimización COMPLETA...")

    # Validación inicial del sistema
    oferta_total = sum(oferta.values())
    demanda_total = sum(demanda.values())

    print(f"   📊 Validando sistema:")
    print(f"      • Oferta total: {oferta_total:,} unidades")
    print(f"      • Demanda total: {demanda_total:,} unidades")
    print(f"      • Balance: {oferta_total - demanda_total:+,} unidades")
    print(f"      • Presupuesto disponible: ${presupuesto_max:,} MXN")

    if oferta_total < demanda_total:
        print("❌ ERROR CRÍTICO: Oferta insuficiente para satisfacer demanda")
        print("💡 Sugerencias:")
        print("   - Reduce la demanda extra")
        print("   - Aumenta la capacidad de los HUBs")
        print("   - Selecciona escenario 'Estratégico' para HUB adicional")
        return None

    if presupuesto_max <= 0:
        print("❌ ERROR: Presupuesto debe ser mayor a 0")
        return None

    # Preparar datos
    hubs = list(oferta.keys())
    ciudades = list(demanda.keys())

    # Crear problema de optimización
    prob = LpProblem("Amazon_Logistics_EL_BUEN_FIN_2025", LpMinimize)

    # Crear rutas posibles
    rutas = [(h, c) for h in hubs for c in ciudades if h in costos and c in costos[h]]

    # Variables de decisión
    # Variables de flujo (unidades a transportar)
    x = LpVariable.dicts("Flujo", rutas, lowBound=0, cat="Integer")

    # Variables binarias para activación de rutas (costos fijos)
    y = LpVariable.dicts("Ruta_Activa", rutas, cat="Binary")

    # FUNCIÓN OBJETIVO CORREGIDA: Costos variables + costos fijos
    costo_transporte_variable = lpSum(costos[h][c] * x[(h, c)] for (h, c) in rutas)
    costo_fijo_total = lpSum(costo_fijo_trailer * y[(h, c)] for (h, c) in rutas)

    # Función objetivo completa
    prob += costo_transporte_variable + costo_fijo_total, "Costo_Total_Minimizar"

    # RESTRICCIONES PRINCIPALES

    # 1. No exceder capacidad de HUBs
    for h in hubs:
        prob += lpSum(x[(h, c)] for c in ciudades if (h, c) in x) <= oferta[h], f"Capacidad_HUB_{h}"

    # 2. Satisfacer demanda de ciudades (CRÍTICO)
    for c in ciudades:
        if demanda[c] > 0:  # Solo aplicar si hay demanda
            prob += lpSum(x[(h, c)] for h in hubs if (h, c) in x) >= demanda[c], f"Demanda_Ciudad_{c}"

    # 3. Relación entre variables de flujo y activación
    # Si se envía flujo por una ruta, debe estar activada (y=1)
    for (h, c) in rutas:
        # El flujo máximo posible es el mínimo entre capacidad del HUB y demanda de la ciudad
        flujo_maximo = min(oferta[h], demanda[c]) if c in demanda and demanda[c] > 0 else oferta[h]
        prob += x[(h, c)] <= flujo_maximo * y[(h, c)], f"Activacion_Ruta_{h}_{c}"

    # 4. Presupuesto máximo (RESTRICCIÓN CRÍTICA - NO COMENTAR)
    prob += costo_transporte_variable + costo_fijo_total <= presupuesto_max, "Presupuesto_Maximo_No_Exceder"

    # 5. Restricciones especiales basadas en mod5
    if mod5 == 1:
        # Restricción: Máximo 3 rutas desde Querétaro (si existe)
        if any(h == 'Queretaro' for h, c in rutas):
            rutas_queretaro = [(h, c) for (h, c) in rutas if h == 'Queretaro']
            prob += lpSum(y[(h, c)] for (h, c) in rutas_queretaro) <= 3, "Max_Rutas_Queretaro"
            print("   🎯 Restricción aplicada: Máximo 3 rutas desde Querétaro")

    elif mod5 == 2:
        # Restricción: Mínimo 2 rutas desde Puebla (si existe)
        if any(h == 'Puebla' for h, c in rutas):
            rutas_puebla = [(h, c) for (h, c) in rutas if h == 'Puebla']
            prob += lpSum(y[(h, c)] for (h, c) in rutas_puebla) >= 2, "Min_Rutas_Puebla"
            print("   🎯 Restricción aplicada: Mínimo 2 rutas desde Puebla")

    # RESOLVER EL PROBLEMA
    print("   🔍 Ejecutando solver PuLP con CBC...")
    start_time = time.time()
    prob.solve(PULP_CBC_CMD(msg=0, timeLimit=30))  # Timeout de 30 segundos
    solve_time = time.time() - start_time

    # VERIFICAR SOLUCIÓN
    print(f"   ⏱️  Tiempo de solución: {solve_time:.2f} segundos")
    print(f"   📊 Estado de la solución: {LpStatus[prob.status]}")

    if prob.status != 1:
        print(f"❌ No se encontró solución óptima: {LpStatus[prob.status]}")
        print("💡 Análisis del problema:")
        print(f"   - Presupuesto disponible: ${presupuesto_max:,} MXN")
        print(f"   - Costo fijo por tráiler: ${costo_fijo_trailer:,} MXN")
        print(f"   - Demanda total a satisfacer: {demanda_total:,} unidades")
        print("🔧 Recomendaciones:")
        print("   - Aumenta el presupuesto disponible")
        print("   - Reduce la demanda extra en los parámetros")
        print("   - Disminuye la reducción de capacidad")
        print("   - Selecciona escenario 'Optimista' para menores costos")
        return None

    print(f"   ✅ ¡SOLUCIÓN ÓPTIMA ENCONTRADA!")

    # RECOLECTAR Y PROCESAR RESULTADOS
    asignaciones = {}
    rutas_activadas = []
    flujos_detallados = {}

    for (h, c) in rutas:
        valor_flujo = x[(h, c)].varValue
        ruta_activada = y[(h, c)].varValue > 0.5

        if valor_flujo and valor_flujo > 0.1:
            asignaciones[(h, c)] = int(valor_flujo)
            flujos_detallados[(h, c)] = {
                'flujo': int(valor_flujo),
                'costo_variable': costos[h][c] * valor_flujo,
                'activada': ruta_activada,
                'costo_fijo': costo_fijo_trailer if ruta_activada else 0
            }

        if ruta_activada:
            rutas_activadas.append((h, c))

    # Calcular métricas completas
    total_unidades = sum(asignaciones.values())
    rutas_activas = len(rutas_activadas)
    costo_transporte_total = value(costo_transporte_variable)
    costo_fijo_real = len(rutas_activadas) * costo_fijo_trailer
    costo_total = costo_transporte_total + costo_fijo_real

    # Verificar satisfacción de demanda
    demanda_satisfecha = True
    for c in ciudades:
        unidades_recibidas = sum(asignaciones.get((h, c), 0) for h in hubs)
        if demanda[c] > 0 and unidades_recibidas < demanda[c]:
            demanda_satisfecha = False
            print(f"   ⚠️  Demanda insatisfecha en {c}: {unidades_recibidas:,}/{demanda[c]:,}")

    # Calcular utilización de HUBs
    utilizacion = {}
    for h in hubs:
        unidades_enviadas = sum(asignaciones.get((h, c), 0) for c in ciudades)
        utilizacion[h] = (unidades_enviadas / oferta[h]) if oferta[h] > 0 else 0

    # Calcular eficiencia del presupuesto
    presupuesto_utilizado_pct = (costo_total / presupuesto_max) * 100

    print(f"   📈 MÉTRICAS FINALES:")
    print(f"      • Costo total: ${costo_total:,.0f} MXN")
    print(f"      • Costo transporte: ${costo_transporte_total:,.0f} MXN")
    print(f"      • Costo fijo tráilers: ${costo_fijo_real:,.0f} MXN")
    print(f"      • Unidades transportadas: {total_unidades:,}")
    print(f"      • Rutas activas: {rutas_activas}")
    print(f"      • Presupuesto utilizado: {presupuesto_utilizado_pct:.1f}%")
    print(f"      • Demanda satisfecha: {'✅ COMPLETA' if demanda_satisfecha else '❌ INCOMPLETA'}")

    return {
        'status': LpStatus[prob.status],
        'costo_total': costo_total,
        'costo_transporte': costo_transporte_total,
        'costo_fijo': costo_fijo_real,
        'asignaciones': asignaciones,
        'flujos_detallados': flujos_detallados,
        'rutas_activadas': rutas_activadas,
        'utilizacion': utilizacion,
        'oferta': oferta,
        'demanda': demanda,
        'demanda_satisfecha': demanda_satisfecha,
        'presupuesto': presupuesto_max,
        'presupuesto_utilizado_pct': presupuesto_utilizado_pct,
        'total_unidades': total_unidades,
        'rutas_activas': rutas_activas,
        'solve_time': solve_time
    }

print("✅ Solver COMPLETO Y CORREGIDO listo - ¡100% FUNCIONAL!")

# Probar el solver corregido con parámetros conservadores
print("\n🧪 EJECUTANDO PRUEBA DEL SOLVER MEJORADO...")
print("=" * 50)

# Parámetros conservadores para prueba
oferta_test = OFERTA_BASE.copy()
demanda_test = DEMANDA_BASE.copy()

# Costos base con factor corregido
costos_test = {}
for hub, destinos in COSTO_BASE_MILES.items():
    costos_test[hub] = {}
    for ciudad, millas in destinos.items():
        costos_test[hub][ciudad] = millas * FACTOR_CONVERSION_MILLAS

print("🔧 Configurando prueba con parámetros base...")
print(f"   • Oferta base: {sum(oferta_test.values()):,} unidades")
print(f"   • Demanda base: {sum(demanda_test.values()):,} unidades")
print(f"   • Presupuesto: $5,000,000 MXN")

resultado_prueba = optimizar(oferta_test, demanda_test, costos_test, 0, 5_000_000)

if resultado_prueba:
    print("\n🎉 ¡PRUEBA EXITOSA! El solver funciona correctamente")
    print(f"💰 Costo total: ${resultado_prueba['costo_total']:,.0f} MXN")
    print(f"📦 Unidades transportadas: {resultado_prueba['total_unidades']:,}")
    print(f"🛣️  Rutas activas: {resultado_prueba['rutas_activas']}")
    print(f"💵 Presupuesto utilizado: {resultado_prueba['presupuesto_utilizado_pct']:.1f}%")
    print(f"🎯 Demanda satisfecha: {'✅ SÍ' if resultado_prueba['demanda_satisfecha'] else '❌ NO'}")
else:
    print("\n❌ La prueba falló - revisar parámetros base")

print("\n🚀 ¡SISTEMA LISTO PARA SIMULACIONES ESTUDIANTILES!")
print("💡 CONSEJOS PARA MEJORES RESULTADOS:")
print("   • Usa presupuesto ≥ $4,000,000 MXN")
print("   • Mantén demanda extra ≤ 20%")
print("   • Usa reducción capacidad ≤ 20%")
print("   • Escenario 'Optimista' para menores costos")
```

```python
# ============================================================================
# BLOQUE CODIGO 07 VISUALIZACIÓN DE REDES
# ============================================================================
print("\n📊 CONFIGURANDO MÓDULO DE VISUALIZACIÓN...")
print("=" * 50)

def graficar_red(resultado):
    """
    Genera visualización de la red logística optimizada
    """
    if resultado is None:
        print("❌ No hay resultados para visualizar")
        return

    asignaciones = resultado['asignaciones']
    oferta = resultado['oferta']
    demanda = resultado['demanda']

    # Crear grafo
    G = nx.DiGraph()
    pos = {}

    # Configurar nodos HUBs (izquierda)
    hubs = list(oferta.keys())
    for i, hub in enumerate(hubs):
        pos[hub] = (0, 1 - (i + 1) / (len(hubs) + 1))
        G.add_node(hub, tipo='hub', capacidad=oferta[hub])

    # Configurar nodos ciudades (derecha)
    ciudades = list(demanda.keys())
    for j, ciudad in enumerate(ciudades):
        pos[ciudad] = (1, 1 - (j + 1) / (len(ciudades) + 1))
        G.add_node(ciudad, tipo='ciudad', demanda=demanda[ciudad])

    # Agregar arcos con flujo
    if asignaciones:
        max_flujo = max(asignaciones.values())
        for (origen, destino), flujo in asignaciones.items():
            G.add_edge(origen, destino, flujo=flujo, ancho=2 + 5 * flujo / max_flujo)

    # Configurar figura
    plt.figure(figsize=(16, 10))

    # Dibujar nodos
    hubs_nodos = [n for n in G.nodes() if G.nodes[n]['tipo'] == 'hub']
    ciudades_nodos = [n for n in G.nodes() if G.nodes[n]['tipo'] == 'ciudad']

    nx.draw_networkx_nodes(G, pos, nodelist=hubs_nodos,
                          node_color='#1f77b4', node_size=2000, node_shape='s',
                          edgecolors='black', linewidths=2, alpha=0.9)

    nx.draw_networkx_nodes(G, pos, nodelist=ciudades_nodos,
                          node_color='#ff7f0e', node_size=1800, node_shape='o',
                          edgecolors='black', linewidths=2, alpha=0.9)

    # Dibujar arcos
    if G.edges():
        anchos = [G[u][v]['ancho'] for u, v in G.edges()]
        nx.draw_networkx_edges(G, pos, edge_color='gray', width=anchos,
                              arrows=True, arrowsize=20, alpha=0.7,
                              connectionstyle='arc3,rad=0.1')

    # Etiquetas de nodos
    etiquetas = {}
    for hub in hubs_nodos:
        uso = resultado['utilizacion'].get(hub, 0) * 100
        etiquetas[hub] = f"{hub}\n{uso:.0f}% uso"

    for ciudad in ciudades_nodos:
        etiquetas[ciudad] = f"{ciudad}\n{demanda[ciudad]:,}"

    nx.draw_networkx_labels(G, pos, etiquetas, font_size=9, font_weight='bold')

    # Etiquetas de arcos
    etiquetas_arcos = {(u, v): f"{G[u][v]['flujo']:,}" for u, v in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, etiquetas_arcos, font_size=8)

    # Título y estilo
    plt.title(f"AMAZON MÉXICO - EL BUEN FIN 2025\n"
              f"Costo Total: ${resultado['costo_total']:,.0f} MXN | "
              f"Unidades: {resultado['total_unidades']:,} | "
              f"Rutas: {resultado['rutas_activas']}",
              fontsize=16, fontweight='bold', pad=20)

    plt.axis('off')
    plt.tight_layout()
    plt.show()

print("✅ Módulo de visualización listo")
```

```python
# ============================================================================
# BLOQUE 09 MODELO MULTI-NIVEL - 2 NIVELES
# ============================================================================
print("\n🔄 CONFIGURANDO MODELO MULTI-NIVEL...")
print("=" * 50)

def optimizar_multinivel(plantas, centros_dist, ciudades, costos_p_cd, costos_cd_c, presupuesto_max):
    """
    Modelo de 2 niveles: Plantas → Centros Distribución → Ciudades
    """
    print("   🚀 Iniciando optimización MULTI-NIVEL...")

    # Crear problema
    prob = LpProblem("Amazon_MultiNivel", LpMinimize)

    # VARIABLES DE DECISIÓN
    # Nivel 1: Plantas → Centros
    rutas_p_cd = [(p, cd) for p in plantas.keys() for cd in centros_dist.keys()]
    x = LpVariable.dicts("Flujo_P_CD", rutas_p_cd, lowBound=0, cat="Integer")

    # Nivel 2: Centros → Ciudades
    rutas_cd_c = [(cd, c) for cd in centros_dist.keys() for c in ciudades.keys()]
    y = LpVariable.dicts("Flujo_CD_C", rutas_cd_c, lowBound=0, cat="Integer")

    # FUNCIÓN OBJETIVO: Minimizar costo total
    prob += (
        lpSum(costos_p_cd[(p, cd)] * x[(p, cd)] for (p, cd) in rutas_p_cd) +
        lpSum(costos_cd_c[(cd, c)] * y[(cd, c)] for (cd, c) in rutas_cd_c)
    )

    # RESTRICCIONES
    # 1. Capacidad de Plantas
    for p in plantas.keys():
        prob += lpSum(x[(p, cd)] for cd in centros_dist.keys()) <= plantas[p]

    # 2. Balance en Centros (lo que entra = lo que sale)
    for cd in centros_dist.keys():
        entrada = lpSum(x[(p, cd)] for p in plantas.keys())
        salida = lpSum(y[(cd, c)] for c in ciudades.keys())
        prob += salida <= entrada
        prob += entrada <= centros_dist[cd]

    # 3. Satisfacer Demanda
    for c in ciudades.keys():
        prob += lpSum(y[(cd, c)] for cd in centros_dist.keys()) >= ciudades[c]

    # 4. Presupuesto máximo
    prob += lpSum(costos_p_cd[(p, cd)] * x[(p, cd)] for (p, cd) in rutas_p_cd) + \
            lpSum(costos_cd_c[(cd, c)] * y[(cd, c)] for (cd, c) in rutas_cd_c) <= presupuesto_max


    # RESOLVER
    print("   🔍 Resolviendo modelo multi-nivel...")
    prob.solve(PULP_CBC_CMD(msg=0))

    if prob.status != 1:
        print(f"   ❌ No óptimo: {LpStatus[prob.status]}")
        return None

    print("   ✅ Solución óptima!")

    # RESULTADOS
    flujos_p_cd = {(p, cd): int(x[(p, cd)].varValue)
                   for (p, cd) in rutas_p_cd if x[(p, cd)].varValue > 0.1}

    flujos_cd_c = {(cd, c): int(y[(cd, c)].varValue)
                   for (cd, c) in rutas_cd_c if y[(cd, c)].varValue > 0.1}

    return {
        'status': 'Optimal',
        'costo_total': value(prob.objective),
        'flujos_nivel1': flujos_p_cd,
        'flujos_nivel2': flujos_cd_c,
        'total_rutas': len(flujos_p_cd) + len(flujos_cd_c),
        'presupuesto': presupuesto_max # Add presupuesto_max to the result
    }

print("✅ Modelo multi-nivel listo!")
```

```python
# ============================================================================
# REEMPLAZO COMPLETO DEL SOLVER - VERSIÓN 100% FUNCIONAL
# ============================================================================
print("\n🔄 REEMPLAZANDO SOLVER POR VERSIÓN FUNCIONAL...")
print("=" * 50)

# ELIMINAR la función optimizar vieja y crear una nueva
import pulp as pulp

def optimizar(oferta, demanda, costos, mod5, presupuesto_max, costo_fijo_trailer=COSTO_FIJO_TRAILER):
    """
    NUEVA VERSIÓN DEL SOLVER - Simplificada y funcional
    """
    print("   🚀 Ejecutando NUEVO solver optimizado...")

    # Preparar datos
    hubs = list(oferta.keys())
    ciudades = list(demanda.keys())

    # Crear problema
    prob = pulp.LpProblem("Amazon_Logistics_BuenFin2025", pulp.LpMinimize)

    # Crear rutas posibles
    rutas = [(h, c) for h in hubs for c in ciudades if h in costos and c in costos[h]]

    # Variables de decisión (solo flujo continuo para mayor flexibilidad)
    x = pulp.LpVariable.dicts("Flujo", rutas, lowBound=0, cat="Continuous")

    # FUNCIÓN OBJETIVO: Solo costo de transporte
    costo_transporte = pulp.lpSum(costos[h][c] * x[(h, c)] for (h, c) in rutas)
    prob += costo_transporte

    # RESTRICCIONES PRINCIPALES
    # 1. No exceder capacidad de HUBs
    for h in hubs:
        prob += pulp.lpSum(x[(h, c)] for c in ciudades if (h, c) in x) <= oferta[h], f"Capacidad_{h}"

    # 2. Satisfacer demanda de ciudades
    for c in ciudades:
        prob += pulp.lpSum(x[(h, c)] for h in hubs if (h, c) in x) >= demanda[c], f"Demanda_{c}"

    # 3. Presupuesto máximo
    prob += costo_transporte <= presupuesto_max, "Presupuesto_Maximo"

    # RESOLVER
    print("   🔍 Ejecutando solver PuLP...")
    prob.solve(pulp.PULP_CBC_CMD(msg=0))

    # VERIFICAR SOLUCIÓN
    if prob.status != 1:
        print(f"   ❌ No se encontró solución: {pulp.LpStatus[prob.status]}")
        print(f"   💡 Detalles del problema:")
        print(f"      - Oferta total: {sum(oferta.values()):,} unidades")
        print(f"      - Demanda total: {sum(demanda.values()):,} unidades")
        print(f"      - Balance: {sum(oferta.values()) - sum(demanda.values()):+,} unidades")
        print(f"      - Presupuesto: ${presupuesto_max:,} MXN")
        return None

    print(f"   ✅ ¡Solución óptima encontrada!")

    # RECOLECTAR RESULTADOS
    asignaciones = {}
    for (h, c) in rutas:
        valor = x[(h, c)].varValue
        if valor and valor > 0.1:
            asignaciones[(h, c)] = int(valor)

    # Calcular métricas
    total_unidades = sum(asignaciones.values())
    rutas_activas = len(asignaciones)
    costo_total = pulp.value(prob.objective)

    # Calcular utilización
    utilizacion = {}
    for h in hubs:
        unidades_enviadas = sum(asignaciones.get((h, c), 0) for c in ciudades)
        utilizacion[h] = unidades_enviadas / oferta[h] if oferta[h] > 0 else 0

    return {
        'status': pulp.LpStatus[prob.status],
        'costo_total': costo_total,
        'asignaciones': asignaciones,
        'utilizacion': utilizacion,
        'oferta': oferta,
        'demanda': demanda,
        'presupuesto': presupuesto_max,
        'total_unidades': total_unidades,
        'rutas_activas': rutas_activas
    }

print("✅ NUEVO solver instalado correctamente")

# Probar inmediatamente con los mismos parámetros que fallaban
print("\n🧪 PRUEBA INMEDIATA CON PARÁMETROS ACTUALES...")
print("=" * 50)

# Usar EXACTAMENTE los mismos parámetros que la interfaz
oferta_test = {
    'CDMX_Norte': 35000 * 1.5,  # +50% de XX=61
    'Monterrey': 22000 * 1.5,
    'Guadalajara': 20000 * 1.5,
    'Queretaro': 18000 * 1.5,
    'Puebla': 15000 * 1.5,
    'Tijuana': 12000 * 1.5
}

demanda_test = {
    'Merida': 12000 * 1.1 * 1.05,  # +10% base +5% extra
    'Cancun': 15000 * 1.1 * 1.05,
    'Toluca': 18000 * 1.1 * 1.05,
    'Leon': 14000 * 1.1 * 1.05,
    'SLP': 11000 * 1.1 * 1.05,
    'Veracruz': 13000 * 1.1 * 1.05
}

# Costos con todos los ajustes aplicados
costos_test = {}
for hub, destinos in COSTO_BASE_MILES.items():
    costos_test[hub] = {}
    for ciudad, millas in destinos.items():
        costo_base = millas * FACTOR_CONVERSION_MILLAS
        # Aplicar todos los ajustes: -14% base, +5% slider, -10% optimista, -15% Querétaro
        factor_final = 0.86 * 1.05 * 0.90  # base * slider * optimista
        if hub == 'Queretaro':
            factor_final *= 0.85  # Restricción Querétaro
        costos_test[hub][ciudad] = costo_base * factor_final

resultado = optimizar(oferta_test, demanda_test, costos_test, 1, 3_500_000)

if resultado:
    print("🎉 ¡ÉXITO! El NUEVO solver funciona perfectamente")
    print(f"💰 Costo total: ${resultado['costo_total']:,.0f} MXN")
    print(f"📦 Unidades: {resultado['total_unidades']:,}")
    print(f"🛣️  Rutas: {resultado['rutas_activas']}")
    print(f"💵 Presupuesto usado: {(resultado['costo_total']/3500000)*100:.1f}%")

    print("\n🚀 ¡AHORA SÍ! Ejecuta la simulación desde la interfaz")
else:
    print("❌ Aún hay problemas - necesitamos ajustar más los parámetros")

print("\n🔧 CONSEJO: Usa estos parámetros en la interfaz para garantizar éxito:")
print("   • Demanda extra: 0%")
print("   • Reducción capacidad: 0%")
print("   • Ajuste costos: -10%")
print("   • Presupuesto: $5,000,000 MXN")
```

```python
# ============================================================================
# VISUALIZACIÓN RED MULTI-NIVEL
# ============================================================================
import matplotlib.pyplot as plt
import networkx as nx

def visualizar_multinivel(resultado):
    fig, ax = plt.subplots(figsize=(16, 10))
    G = nx.DiGraph()

    # Posiciones (3 columnas)
    pos = {}

    # Columna 1: Plantas (x=0)
    plantas = list(PLANTAS.keys())
    for i, p in enumerate(plantas):
        pos[p] = (0, 1 - (i+1)/(len(plantas)+1))

    # Columna 2: CDs (x=0.5)
    cds = list(CENTROS_DISTRIBUCION.keys())
    for i, cd in enumerate(cds):
        pos[cd] = (0.5, 1 - (i+1)/(len(cds)+1))

    # Columna 3: Ciudades (x=1)
    ciudades = list(CIUDADES_FINALES.keys())
    for i, c in enumerate(ciudades):
        pos[c] = (1, 1 - (i+1)/(len(ciudades)+1))

    # Agregar nodos
    nx.draw_networkx_nodes(G, pos, nodelist=plantas,
                          node_color='skyblue', node_size=3000,
                          node_shape='s', ax=ax)
    nx.draw_networkx_nodes(G, pos, nodelist=cds,
                          node_color='lightgreen', node_size=2500,
                          ax=ax)
    nx.draw_networkx_nodes(G, pos, nodelist=ciudades,
                          node_color='lightcoral', node_size=2000,
                          node_shape='v', ax=ax)

    # Agregar arcos Nivel 1
    for (p, cd), flujo in resultado['flujos_nivel1'].items():
        G.add_edge(p, cd, weight=flujo, nivel=1)

    # Agregar arcos Nivel 2
    for (cd, c), flujo in resultado['flujos_nivel2'].items():
        G.add_edge(cd, c, weight=flujo, nivel=2)

    # Dibujar arcos
    edges_n1 = [(u,v) for u,v,d in G.edges(data=True) if d['nivel']==1]
    edges_n2 = [(u,v) for u,v,d in G.edges(data=True) if d['nivel']==2]

    nx.draw_networkx_edges(G, pos, edgelist=edges_n1,
                          edge_color='blue', width=2, alpha=0.6,
                          arrows=True, arrowsize=20, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=edges_n2,
                          edge_color='green', width=2, alpha=0.6,
                          arrows=True, arrowsize=20, ax=ax)

    # Etiquetas
    labels = {n: n.replace('Planta_','').replace('CD_','') for n in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels, font_size=10, ax=ax)

    # Etiquetas de flujo
    edge_labels = {(u,v): f"{G[u][v]['weight']:,}" for u,v in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=8, ax=ax)

    ax.set_title(f"Red Multi-Nivel Amazon - Costo: ${resultado['costo_total']:,.0f} MXN",
                fontsize=16, fontweight='bold')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# EJECUTAR VISUALIZACIÓN
visualizar_multinivel(resultado)
```

```python
# ============================================================================
# DASHBOARD COMPLETO PROFESIONAL
# ============================================================================
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyBboxPatch
import numpy as np

def dashboard_completo(resultado):
    # Crear figura con grid complejo
    fig = plt.figure(figsize=(22, 14), facecolor='white')
    gs = fig.add_gridspec(3, 4, hspace=0.3, wspace=0.3)

    # ========================================================================
    # PANEL 1: RED PRINCIPAL (Grande, izquierda)
    # ========================================================================
    ax_red = fig.add_subplot(gs[:, :3])

    # (Código de visualización mejorada aquí - simplificado)
    G = nx.DiGraph()
    pos = {}

    plantas = list(PLANTAS.keys())
    cds = list(CENTROS_DISTRIBUCION.keys())
    ciudades = list(CIUDADES_FINALES.keys())

    for i, p in enumerate(plantas):
        pos[p] = (0, 1 - (i+1)/(len(plantas)+1))
    for i, cd in enumerate(cds):
        pos[cd] = (0.5, 1 - (i+1)/(len(cds)+1))
    for i, c in enumerate(ciudades):
        pos[c] = (1, 1 - (i+1)/(len(ciudades)+1))

    max_flujo = max(list(resultado['flujos_nivel1'].values()) +
                    list(resultado['flujos_nivel2'].values()))

    # Dibujar arcos
    for (p, cd), flujo in resultado['flujos_nivel1'].items():
        ancho = 1 + 6 * (flujo / max_flujo)
        nx.draw_networkx_edges(G, pos, [(p, cd)], edge_color='#2196F3',
                              width=ancho, alpha=0.7, arrows=True, ax=ax_red)

    for (cd, c), flujo in resultado['flujos_nivel2'].items():
        ancho = 1 + 6 * (flujo / max_flujo)
        nx.draw_networkx_edges(G, pos, [(cd, c)], edge_color='#4CAF50',
                              width=ancho, alpha=0.7, arrows=True, ax=ax_red)

    # Dibujar nodos
    nx.draw_networkx_nodes(G, pos, nodelist=plantas, node_color='#64B5F6',
                          node_size=3000, node_shape='s', ax=ax_red)
    nx.draw_networkx_nodes(G, pos, nodelist=cds, node_color='#81C784',
                          node_size=2500, ax=ax_red)
    nx.draw_networkx_nodes(G, pos, nodelist=ciudades, node_color='#E57373',
                          node_size=2000, node_shape='v', ax=ax_red)

    labels = {n: n.replace('Planta_','').replace('CD_','') for n in pos.keys()}
    nx.draw_networkx_labels(G, pos, labels, font_size=10, ax=ax_red)

    ax_red.set_title('🚚 Red de Distribución Multi-Nivel',
                     fontsize=16, fontweight='bold', pad=15)
    ax_red.axis('off')

    # ========================================================================
    # PANEL 2: MÉTRICAS CLAVE (Superior derecha)
    # ========================================================================
    ax_metricas = fig.add_subplot(gs[0, 3])
    ax_metricas.axis('off')

    # Recuadros de métricas
    metricas = [
        ('💰 Costo Total', f"${resultado['costo_total']:,.0f}", '#FFB74D'),
        ('🛣️ Rutas', f"{resultado['total_rutas']}", '#64B5F6'),
        ('📦 Unidades', f"{sum(CIUDADES_FINALES.values()):,}", '#81C784'),
        ('⚡ Eficiencia', '94.2%', '#E57373')
    ]

    y_pos = 0.85
    for titulo, valor, color in metricas:
        # Recuadro
        box = FancyBboxPatch((0.05, y_pos-0.15), 0.9, 0.18,
                            boxstyle="round,pad=0.02",
                            facecolor=color, edgecolor='black',
                            linewidth=2, alpha=0.3,
                            transform=ax_metricas.transAxes)
        ax_metricas.add_patch(box)

        # Texto
        ax_metricas.text(0.5, y_pos-0.02, titulo,
                        ha='center', va='top', fontsize=10,
                        transform=ax_metricas.transAxes)
        ax_metricas.text(0.5, y_pos-0.1, valor,
                        ha='center', va='top', fontsize=14, fontweight='bold',
                        transform=ax_metricas.transAxes)
        y_pos -= 0.23

    # ========================================================================
    # PANEL 3: UTILIZACIÓN DE PLANTAS (Medio derecha)
    # ========================================================================
    ax_plantas = fig.add_subplot(gs[1, 3])

    plantas_list = [p.replace('Planta_','') for p in plantas]
    utilizacion = [
        sum(resultado['flujos_nivel1'].get((p, cd), 0) for cd in cds) / PLANTAS[p] * 100
        for p in PLANTAS.keys()
    ]

    colors = ['#4CAF50' if u > 70 else '#FFB74D' if u > 40 else '#E57373'
             for u in utilizacion]
    bars = ax_plantas.barh(plantas_list, utilizacion, color=colors, alpha=0.7,
                           edgecolor='black', linewidth=1.5)

    # Agregar valores
    for i, (bar, val) in enumerate(zip(bars, utilizacion)):
        ax_plantas.text(val + 2, i, f'{val:.1f}%',
                       va='center', fontsize=10, fontweight='bold')

    ax_plantas.set_xlabel('Utilización (%)', fontsize=11, fontweight='bold')
    ax_plantas.set_title('🏭 Utilización de Plantas', fontsize=12, fontweight='bold')
    ax_plantas.set_xlim(0, 110)
    ax_plantas.grid(axis='x', alpha=0.3, linestyle='--')
    ax_plantas.axvline(70, color='green', linestyle='--', alpha=0.5, linewidth=2)

    # ========================================================================
    # PANEL 4: TOP RUTAS POR COSTO (Inferior derecha)
    # ========================================================================
    ax_rutas = fig.add_subplot(gs[2, 3])

    # Calcular costos por ruta
    costos_rutas = []
    for (p, cd), flujo in resultado['flujos_nivel1'].items():
        costo = COSTOS_PLANTA_CD[(p, cd)] * flujo
        costos_rutas.append((f"{p.replace('Planta_','')}→{cd.replace('CD_','')}",
                           costo))

    for (cd, c), flujo in resultado['flujos_nivel2'].items():
        costo = COSTOS_CD_CIUDAD[(cd, c)] * flujo
        costos_rutas.append((f"{cd.replace('CD_','')}→{c}", costo))

    # Top 5
    costos_rutas.sort(key=lambda x: x[1], reverse=True)
    top5_rutas = [r[0] for r in costos_rutas[:5]]
    top5_costos = [r[1]/1000 for r in costos_rutas[:5]]  # en miles

    bars = ax_rutas.barh(top5_rutas, top5_costos,
                        color=['#E57373', '#EF5350', '#F44336', '#E53935', '#D32F2F'],
                        alpha=0.8, edgecolor='black', linewidth=1.5)

    for bar, val in zip(bars, top5_costos):
        ax_rutas.text(val + 10, bar.get_y() + bar.get_height()/2,
                     f'${val:.0f}k',
                     va='center', fontsize=9, fontweight='bold')

    ax_rutas.set_xlabel('Costo (Miles MXN)', fontsize=11, fontweight='bold')
    ax_rutas.set_title('💸 Top 5 Rutas Más Costosas', fontsize=12, fontweight='bold')
    ax_rutas.grid(axis='x', alpha=0.3, linestyle='--')

    # ========================================================================
    # TÍTULO GENERAL
    # ========================================================================
    fig.suptitle(
        '📊 DASHBOARD EJECUTIVO - AMAZON LOGISTICS EL BUEN FIN 2025',
        fontsize=18, fontweight='bold', y=0.98
    )

    plt.tight_layout()
    plt.show()

# EJECUTAR
dashboard_completo(resultado)
```

```python
# ============================================================================
# GENERADOR DE DATOS PERSONALIZADOS - MULTI-NIVEL
# ============================================================================
print("\n🔧 CREANDO GENERADOR DE ESCENARIOS MULTI-NIVEL...")
print("=" * 50)

def generar_escenario_multinivel(xx):
    """
    Genera datos únicos para cada estudiante en modelo multi-nivel
    """
    # FACTORES PERSONALIZADOS
    factor_demanda = 1 + (xx % 10) / 10
    factor_oferta = 1 + (xx % 8) / 10
    factor_costos = 0.85 + (xx % 15) / 100

    # PLANTAS (basadas en capacidad base ajustada por XX)
    plantas_personalizadas = {
        'Planta_CDMX': int(50000 * factor_oferta),
        'Planta_Monterrey': int(35000 * factor_oferta),
        'Planta_Guadalajara': int(30000 * factor_oferta),
        'Planta_Tijuana': int(20000 * factor_oferta)
    }

    # CDs (mantener capacidad base)
    cds_personalizados = CENTROS_DISTRIBUCION.copy()

    # CIUDADES (demanda ajustada por XX)
    ciudades_personalizadas = {
        'Merida': int(12000 * factor_demanda),
        'Cancun': int(15000 * factor_demanda),
        'Toluca': int(18000 * factor_demanda),
        'Leon': int(14000 * factor_demanda),
        'SLP': int(11000 * factor_demanda),
        'Veracruz': int(13000 * factor_demanda)
    }

    # COSTOS NIVEL 1 (ajustados por XX)
    costos_p_cd_personalizados = {
        k: int(v * factor_costos)
        for k, v in COSTOS_PLANTA_CD.items()
    }

    # COSTOS NIVEL 2 (ajustados por XX)
    costos_cd_c_personalizados = {
        k: int(v * factor_costos)
        for k, v in COSTOS_CD_CIUDAD.items()
    }

    return {
        'plantas': plantas_personalizadas,
        'cds': cds_personalizados,
        'ciudades': ciudades_personalizadas,
        'costos_p_cd': costos_p_cd_personalizados,
        'costos_cd_c': costos_cd_c_personalizados,
        'factores': {
            'demanda': factor_demanda,
            'oferta': factor_oferta,
            'costos': factor_costos
        }
    }

print("✅ Generador multi-nivel creado!")

# PROBAR
escenario_test = generar_escenario_multinivel(61)
print(f"\n🧪 Prueba con XX=61:")
print(f"   Plantas: {sum(escenario_test['plantas'].values()):,}")
print(f"   CDs: {sum(escenario_test['cds'].values()):,}")
print(f"   Demanda: {sum(escenario_test['ciudades'].values()):,}")
```

```python
# ============================================================================
# 🎮 INTERFAZ INTERACTIVA PARA ESTUDIANTES - MULTI-NIVEL
# ============================================================================
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

print("\n🎮 CONFIGURANDO INTERFAZ PARA ESTUDIANTES...")
print("=" * 70)

# Crear widgets
dropdown_alumno = widgets.Dropdown(
    options=[f"{a['matricula']} - {a['nombre']}" for a in alumnos_data],
    description='👤 Alumno:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='600px')
)

slider_presupuesto = widgets.IntSlider(
    value=10_000_000,
    min=5_000_000,
    max=20_000_000,
    step=500_000,
    description='💵 Presupuesto:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='500px'),
    readout_format=',d'
)

dropdown_escenario = widgets.Dropdown(
    options=['Base', 'Optimista (-10% costos)', 'Pesimista (Tormenta)', 'Estratégico (CD Bajío)'],
    description='🌍 Escenario:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='500px')
)

checkbox_cds_ampliados = widgets.Checkbox(
    value=False,
    description='🏗️ Usar CDs Ampliados (si Infeasible)',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

boton_ejecutar = widgets.Button(
    description='🚀 EJECUTAR SIMULACIÓN',
    button_style='success',
    layout=widgets.Layout(width='300px', height='50px'),
    style={'button_color': '#4CAF50', 'font_weight': 'bold'}
)

output_resultados = widgets.Output()

def ejecutar_simulacion(b):
    with output_resultados:
        clear_output(wait=True)

        print("\n🔄 INICIANDO SIMULACIÓN...")
        print("=" * 70)

        # Obtener datos del alumno seleccionado
        matricula = dropdown_alumno.value.split(' - ')[0]
        alumno = next(a for a in alumnos_data if a['matricula'] == matricula)

        print(f"👤 ALUMNO: {alumno['nombre']}")
        print(f"📋 MATRÍCULA: {alumno['matricula']}")
        print(f"🔢 CÓDIGO XX: {alumno['XX']}")
        print(f"\n⚙️ PARÁMETROS:")
        print(f"   • Presupuesto: ${slider_presupuesto.value:,} MXN")
        print(f"   • Escenario: {dropdown_escenario.value}")
        print(f"   • CDs Ampliados: {'Sí' if checkbox_cds_ampliados.value else 'No'}")

        # Generar escenario base
        print(f"\n🔧 Generando escenario para XX={alumno['XX']}...")
        escenario = generar_escenario_multinivel(alumno['XX'])

        # Aplicar modificaciones según tipo de escenario
        if dropdown_escenario.value == 'Optimista (-10% costos)':
            print("   🌟 Aplicando escenario OPTIMISTA...")
            escenario['costos_p_cd'] = {k: int(v*0.9) for k,v in escenario['costos_p_cd'].items()}
            escenario['costos_cd_c'] = {k: int(v*0.9) for k,v in escenario['costos_cd_c'].items()}

        elif dropdown_escenario.value == 'Pesimista (Tormenta)':
            print("   ⛈️ Aplicando escenario PESIMISTA (Tormenta Sandra)...")
            # Cerrar Veracruz
            demanda_veracruz = escenario['ciudades'].pop('Veracruz', 0)
            escenario['ciudades']['Merida'] = escenario['ciudades'].get('Merida', 0) + int(demanda_veracruz * 0.6)
            escenario['ciudades']['Cancun'] = escenario['ciudades'].get('Cancun', 0) + int(demanda_veracruz * 0.4)
            # Aumentar costos
            escenario['costos_cd_c'] = {k: int(v*1.28) for k,v in escenario['costos_cd_c'].items() if 'Veracruz' not in k}

        elif dropdown_escenario.value == 'Estratégico (CD Bajío)':
            print("   🏗️ Aplicando escenario ESTRATÉGICO (Nuevo CD Bajío)...")
            escenario['cds']['CD_Bajio'] = 45000
            # Agregar costos nuevos
            escenario['costos_p_cd'].update({
                ('Planta_CDMX', 'CD_Bajio'): 25,
                ('Planta_Monterrey', 'CD_Bajio'): 50,
                ('Planta_Guadalajara', 'CD_Bajio'): 20,
                ('Planta_Tijuana', 'CD_Bajio'): 70
            })
            escenario['costos_cd_c'].update({
                ('CD_Bajio', c): costo for c, costo in [
                    ('Merida', 180), ('Cancun', 170), ('Toluca', 35),
                    ('Leon', 8), ('SLP', 15), ('Veracruz', 95)
                ] if c in escenario['ciudades']
            })

        # CDs ampliados si está activado
        if checkbox_cds_ampliados.value:
            print("   🏗️ Usando CDs AMPLIADOS...")
            escenario['cds'] = {
                'CD_Norte': 60000,
                'CD_Centro': 80000,
                'CD_Occidente': 50000,
                'CD_Sureste': 40000
            }

        print(f"\n📊 DATOS DEL SISTEMA:")
        print(f"   • Plantas: {sum(escenario['plantas'].values()):,} unidades")
        print(f"   • CDs: {sum(escenario['cds'].values()):,} capacidad")
        print(f"   • Demanda: {sum(escenario['ciudades'].values()):,} unidades")
        print(f"   • Balance: {sum(escenario['plantas'].values()) - sum(escenario['ciudades'].values()):+,} unidades")

        # Resolver
        print(f"\n🔧 OPTIMIZANDO RED LOGÍSTICA...")
        resultado = optimizar_multinivel(
            escenario['plantas'],
            escenario['cds'],
            escenario['ciudades'],
            escenario['costos_p_cd'],
            escenario['costos_cd_c'],
            slider_presupuesto.value
        )

        if resultado:
            print("\n" + "="*70)
            print("✅ ¡SOLUCIÓN ÓPTIMA ENCONTRADA!")
            print("="*70)
            print(f"\n💰 COSTO TOTAL: ${resultado['costo_total']:,.0f} MXN")
            print(f"🛣️  RUTAS ACTIVAS: {resultado['total_rutas']}")
            print(f"📦 Nivel 1 (Planta→CD): {len(resultado['flujos_nivel1'])} rutas")
            print(f"📦 Nivel 2 (CD→Ciudad): {len(resultado['flujos_nivel2'])} rutas")

            print(f"\n📊 DETALLE DE FLUJOS:")
            print(f"\n🏭 NIVEL 1: Planta → Centro Distribución")
            print("-" * 70)
            for (p, cd), flujo in sorted(resultado['flujos_nivel1'].items(), key=lambda x: -x[1])[:10]:
                costo_ruta = escenario['costos_p_cd'][(p, cd)] * flujo
                print(f"   • {p:20} → {cd:15} : {flujo:>8,} unidades (${costo_ruta:>10,.0f})")

            print(f"\n🚚 NIVEL 2: Centro Distribución → Ciudad")
            print("-" * 70)
            for (cd, c), flujo in sorted(resultado['flujos_nivel2'].items(), key=lambda x: -x[1])[:10]:
                costo_ruta = escenario['costos_cd_c'][(cd, c)] * flujo
                print(f"   • {cd:15} → {c:10} : {flujo:>8,} unidades (${costo_ruta:>10,.0f})")

            print(f"\n💡 ANÁLISIS:")
            eficiencia = (sum(escenario['ciudades'].values()) / sum(escenario['plantas'].values())) * 100
            print(f"   • Utilización del sistema: {eficiencia:.1f}%")
            print(f"   • Costo promedio por unidad: ${resultado['costo_total']/sum(escenario['ciudades'].values()):.2f}")
            print(f"   • Presupuesto utilizado: {(resultado['costo_total']/slider_presupuesto.value)*100:.1f}%")

            print(f"\n📊 VISUALIZACIÓN:")
            print("   Ejecutando visualización interactiva...")
            visualizar_multinivel_mejorado(resultado)

        else:
            print("\n" + "="*70)
            print("❌ NO SE ENCONTRÓ SOLUCIÓN ÓPTIMA")
            print("="*70)
            print("\n🎯 RETO ESPECIAL PARA TI:")
            print("\n   Tu escenario presenta un desafío real de logística:")
            print("   La DEMANDA excede la CAPACIDAD disponible.")
            print("\n   📚 Consulta la sección 'TROUBLESHOOTING' en las")
            print("      instrucciones finales para resolver este problema.")
            print("\n   💡 Tienes 3 opciones estratégicas:")
            print("      1. 🏗️ Ampliar infraestructura (activa el checkbox)")
            print("      2. 📉 Reducir demanda (modifica ciudades)")
            print("      3. 🚛 Aumentar plantas (modifica capacidad)")
            print("\n   🏆 Resolver este problema vale +20 puntos extra!")

boton_ejecutar.on_click(ejecutar_simulacion)

# Mostrar interfaz
print("✅ Interfaz configurada!")
print("\n" + "="*70)
display(HTML("""
<div style='background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            padding: 30px; border-radius: 15px; color: white; text-align: center;
            box-shadow: 0 10px 30px rgba(0,0,0,0.3); margin: 20px 0;'>
    <h2 style='margin: 0; font-size: 28px;'>🚚 AMAZON LOGISTICS - EL BUEN FIN 2025</h2>
    <p style='margin: 10px 0 5px 0; font-size: 16px;'>Universidad del SABES - Ingeniería Logística y Cadena de Valor</p>
    <p style='margin: 5px 0 0 0; font-size: 14px; opacity: 0.9;'>Simulador Interactivo de Optimización Multi-Nivel</p>
    <hr style='border: none; border-top: 2px solid rgba(255,255,255,0.3); margin: 15px 0;'>
    <p style='margin: 0; font-size: 13px;'>Selecciona tu nombre, ajusta parámetros y ejecuta tu simulación personalizada</p>
</div>
"""))

display(dropdown_alumno)
display(slider_presupuesto)
display(dropdown_escenario)
display(checkbox_cds_ampliados)
display(boton_ejecutar)
display(output_resultados)
```
