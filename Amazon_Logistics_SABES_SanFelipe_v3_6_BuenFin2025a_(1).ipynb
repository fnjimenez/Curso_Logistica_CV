```python
# ============================================================================
# BLOQUE CODIGO 04 CARGA DE DATOS LOG√çSTICOS Y PAR√ÅMETROS FINANCIEROS CORREGIDOS
# ============================================================================
print("\nüè≠ CARGANDO DATOS LOG√çSTICOS Y PAR√ÅMETROS FINANCIEROS...")
print("=" * 50)

# üìä HUBs DE DISTRIBUCI√ìN (OFERTA)
OFERTA_BASE = {
    'CDMX_Norte': 35000,
    'Monterrey':  22000,
    'Guadalajara': 20000,
    'Queretaro':  18000,
    'Puebla':     15000,
    'Tijuana':    12000,
}

print(f"‚úÖ HUBs de distribuci√≥n cargados: {len(OFERTA_BASE)} centros")
print(f"üì¶ Capacidad total: {sum(OFERTA_BASE.values()):,} unidades")

# üéØ CIUDADES DESTINO (DEMANDA)
DEMANDA_BASE = {
    'Merida':   12000,
    'Cancun':   15000,
    'Toluca':   18000,
    'Leon':     14000,
    'SLP':      11000,
    'Veracruz': 13000,
}

print(f"‚úÖ Ciudades destino cargadas: {len(DEMANDA_BASE)} ciudades")
print(f"üéØ Demanda total: {sum(DEMANDA_BASE.values()):,} unidades")

# üîç Validaci√≥n oferta vs demanda
total_oferta = sum(OFERTA_BASE.values())
total_demanda = sum(DEMANDA_BASE.values())
balance = total_oferta - total_demanda

print(f"\nüîç BALANCE OFERTA-DEMANDA:")
print(f"   Oferta total: {total_oferta:,}")
print(f"   Demanda total: {total_demanda:,}")
print(f"   Balance: {total_oferta - total_demanda:+,}")

if balance >= 0:
    print(f"üìä Balance: ‚úÖ Suficiente capacidad (+{balance:,} unidades)")
else:
    print(f"üìä Balance: ‚ùå D√©ficit de {abs(balance):,} unidades")

# üó∫Ô∏è MATRIZ DE COSTOS DE TRANSPORTE (MILES)
COSTO_BASE_MILES = {
    'CDMX_Norte': {'Merida':142, 'Cancun':168, 'Toluca':8, 'Leon':45, 'SLP':52, 'Veracruz':48},
    'Monterrey': {'Merida':228, 'Cancun':245, 'Toluca':88, 'Leon':78, 'SLP':38, 'Veracruz':115},
    'Guadalajara': {'Merida':198, 'Cancun':215, 'Toluca':62, 'Leon':28, 'SLP':58, 'Veracruz':88},
    'Queretaro': {'Merida':158, 'Cancun':178, 'Toluca':18, 'Leon':22, 'SLP':35, 'Veracruz':52},
    'Puebla': {'Merida':128, 'Cancun':145, 'Toluca':22, 'Leon':52, 'SLP':62, 'Veracruz':28},
    'Tijuana': {'Merida':385, 'Cancun':402, 'Toluca':245, 'Leon':235, 'SLP':195, 'Veracruz':272}
}

print(f"‚úÖ Matriz de costos cargada: {len(COSTO_BASE_MILES)}x{len(DEMANDA_BASE)} rutas")

# üîç Validar que todas las rutas existen
rutas_completas = all(len(destinos) == len(DEMANDA_BASE) for destinos in COSTO_BASE_MILES.values())
if rutas_completas:
    print("üîÑ Matriz de costos: ‚úÖ Completa (todas las rutas definidas)")
else:
    print("üîÑ Matriz de costos: ‚ùå Rutas faltantes")

# üí∞ PAR√ÅMETROS FINANCIEROS CORREGIDOS (CR√çTICO)
print("\nüí∞ CONFIGURANDO PAR√ÅMETROS FINANCIEROS REALISTAS...")

# CORRECCI√ìN CR√çTICA: Par√°metros realistas para el modelo
COSTO_FIJO_TRAILER = 12500  # ‚úÖ
FACTOR_CONVERSION_MILLAS = 100  # ‚úÖ Reducido de 1000 a 100
PRESUPUESTO_MAX_DEFAULT = 8_000_000  # ‚úÖ Aumentado a 8M

print("‚úÖ Par√°metros financieros configurados CORRECTAMENTE:")
print(f"   üöõ Costo fijo por tr√°iler: ${COSTO_FIJO_TRAILER:,} MXN")
print(f"   üìä Factor conversi√≥n millas: {FACTOR_CONVERSION_MILLAS}x")
print(f"   üí∞ Presupuesto autorizado: ${PRESUPUESTO_MAX_DEFAULT:,} MXN")

# üîç Validaci√≥n de costos estimados CON FACTOR CORREGIDO
costo_promedio_millas = sum(sum(destinos.values()) for destinos in COSTO_BASE_MILES.values()) / (len(OFERTA_BASE) * len(DEMANDA_BASE))
costo_promedio_mxn = costo_promedio_millas * FACTOR_CONVERSION_MILLAS
print(f"   üìä Costo promedio por ruta: ${costo_promedio_mxn:,.0f} MXN")

print("\nüîç Validaci√≥n de costos por ruta (ejemplos con factor corregido):")
rutas_ejemplo = [
    ('CDMX_Norte', 'Toluca', 8),
    ('CDMX_Norte', 'Merida', 142),
    ('Monterrey', 'Cancun', 245),
    ('Tijuana', 'Veracruz', 272)
]

for origen, destino, millas in rutas_ejemplo:
    costo_estimado = millas * FACTOR_CONVERSION_MILLAS
    print(f"   ‚Ä¢ {origen} ‚Üí {destino}: ${costo_estimado:,} MXN")

print(f"\nüéØ RESUMEN FINAL DEL SISTEMA:")
print(f"   ‚Ä¢ {len(OFERTA_BASE)} HUBs de distribuci√≥n")
print(f"   ‚Ä¢ {len(DEMANDA_BASE)} ciudades destino")
print(f"   ‚Ä¢ {len(OFERTA_BASE) * len(DEMANDA_BASE)} rutas posibles")
print(f"   ‚Ä¢ Balance oferta/demanda: {balance:+,} unidades")
print(f"   ‚Ä¢ Presupuesto disponible: ${PRESUPUESTO_MAX_DEFAULT:,} MXN")
print(f"   ‚Ä¢ Costo fijo por ruta: ${COSTO_FIJO_TRAILER:,} MXN")

print("\nüöÄ ¬°Par√°metros financieros corregidos y listos para optimizaci√≥n!")

# MEJORA RECOMENDADA - Agregar validaci√≥n
def validar_parametros(demanda_extra, capacidad_reduccion, presupuesto):
    if demanda_extra > 0.3:  # M√°ximo 30% de demanda extra
        print("‚ö†Ô∏è  Demanda extra muy alta - puede generar soluciones inviables")
    if capacidad_reduccion > 0.3:  # M√°ximo 30% de reducci√≥n
        print("‚ö†Ô∏è  Reducci√≥n de capacidad muy alta")
    if presupuesto < 3000000:  # Presupuesto m√≠nimo
        print("‚ö†Ô∏è  Presupuesto muy bajo para escenario base")
```

## üé≤ **Paso 5: Generador de Escenario Personalizado** ‚öôÔ∏è

## üéØ **Objetivo del Generador**
Crear **escenarios √∫nicos** para cada estudiante basados en su c√≥digo XX, permitiendo evaluar su capacidad de adaptaci√≥n bajo diferentes condiciones log√≠sticas.

### üîß **Mec√°nica de Personalizaci√≥n**
| Par√°metro | Base | Rango | Factor |
|-----------|------|-------|---------|
| **üìà Demanda** | +0% a +90% | Personalizable | `XX % 10` |
| **üè≠ Oferta** | +0% a +60% | Personalizable | `XX % 7` |
| **üí∞ Costos** | -15% a +14% | Personalizable | `XX % 15` |

### ‚öôÔ∏è **Ajustes Configurables**
- **üéØ Demanda Extra**: Porcentaje adicional sobre base
- **üìâ Reducci√≥n Capacidad**: Limitaci√≥n de recursos
- **üí∏ Ajuste Costos**: Variaci√≥n de precios
- **üè¶ Presupuesto**: L√≠mite financiero personalizado

---

**üöÄ Generador listo para crear desaf√≠os log√≠sticos √∫nicos**

```python
# ============================================================================
# BLOQUE CODIGO 05  GENERADOR DE ESCENARIO PERSONALIZADO (CORREGIDO Y AUDITADO)
# ============================================================================
print("\nüé≤ CONFIGURANDO GENERADOR DE ESCENARIOS...")
print("=" * 50)

def validar_parametros_escenario(demanda_extra, capacidad_reduccion, presupuesto):
    """
    Valida que los par√°metros del escenario sean razonables
    """
    advertencias = []

    if demanda_extra > 0.3:  # M√°ximo 30% de demanda extra
        advertencias.append("‚ö†Ô∏è  Demanda extra > 30% - Alto riesgo de soluci√≥n inviable")

    if capacidad_reduccion > 0.4:  # M√°ximo 40% de reducci√≥n
        advertencias.append("‚ö†Ô∏è  Reducci√≥n capacidad > 40% - Muy restrictivo")

    if presupuesto < 3000000:  # Presupuesto m√≠nimo recommended
        advertencias.append("‚ö†Ô∏è  Presupuesto < $3M - Muy limitado para escenario base")

    if len(advertencias) > 0:
        print("\nüîî ADVERTENCIAS DE PAR√ÅMETROS:")
        for adv in advertencias:
            print(f"   {adv}")
        print("   üí° Recomendaci√≥n: Ajusta los par√°metros para mejor resultado")

    return len(advertencias) == 0  # True si no hay advertencias

def generar_escenario_alumno(XX, demanda_extra_pct, capacidad_reduccion_pct, costo_ajuste_pct, presupuesto_max, escenario_nombre):
    """
    Genera escenario personalizado para el alumno - VERSI√ìN MEJORADA CON VALIDACI√ìN
    """
    print(f"üîß Generando escenario para XX={XX}...")

    # ‚úÖ PRIMERO VALIDAR PAR√ÅMETROS
    parametros_ok = validar_parametros_escenario(demanda_extra_pct, capacidad_reduccion_pct, presupuesto_max)

    if not parametros_ok:
        print("   ‚ö†Ô∏è  Continuando con par√°metros sub√≥ptimos...")

    # C√°lculo de factores
    mod10 = XX % 10
    mod7 = XX % 7
    mod15 = XX % 15
    mod5 = XX % 5

    # Factores base
    factor_demanda = 1 + (mod10 / 10)       # +0% a +90%
    factor_oferta = 1 + (mod7 / 10)         # +0% a +60%
    factor_costos = 0.85 + (mod15 / 100)  # -15% a +14%

    print(f"   üìä Factores base aplicados:")
    print(f"      ‚Ä¢ Demanda: +{(factor_demanda-1)*100:.0f}%")
    print(f"      ‚Ä¢ Oferta: +{(factor_oferta-1)*100:.0f}%")
    print(f"      ‚Ä¢ Costos: {(factor_costos-1)*100:+.1f}%")

    # 1. AJUSTAR DEMANDA
    demanda_ajustada = {}
    for ciudad, base in DEMANDA_BASE.items():
        demanda_ajustada[ciudad] = int(base * factor_demanda * (1 + demanda_extra_pct))

    # 2. AJUSTAR OFERTA
    oferta_ajustada = {}
    for hub, capacidad in OFERTA_BASE.items():
        nueva_capacidad = capacidad * factor_oferta * (1 - capacidad_reduccion_pct)
        oferta_ajustada[hub] = max(5000, int(nueva_capacidad))  # M√≠nimo 5000 unidades

    # 3. AJUSTAR COSTOS (CORRECCI√ìN CR√çTICA APLICADA)
    costos_ajustados = {}
    for hub, destinos in COSTO_BASE_MILES.items():
        costos_ajustados[hub] = {}
        for ciudad, costo_millas in destinos.items():
            # ‚úÖ CORRECCI√ìN: Usar factor realista de conversi√≥n
            costo_mxn = costo_millas * FACTOR_CONVERSION_MILLAS  # 100x in lugar de 1000x
            costo_final = costo_mxn * factor_costos * (1 + costo_ajuste_pct)
            costos_ajustados[hub][ciudad] = costo_final

    # 4. APLICAR ESCENARIO (CORREGIDO)
    if escenario_nombre == "Optimista":
        # Reducir costos 10%
        for hub in costos_ajustados:
            for ciudad in costos_ajustados[hub]:
                costos_ajustados[hub][ciudad] *= 0.90
        print("   üåü Escenario Optimista: Costos -10%")

    elif escenario_nombre == "Pesimista":
        # CORRECCI√ìN CR√çTICA: Tormenta Sandra cierra Veracruz y aumenta costos Sureste (35%)

        # 1. Eliminar la demanda de Veracruz (cierre total)
        if "Veracruz" in demanda_ajustada:
            demanda_ajustada["Veracruz"] = 0

        # 2. Aumentar costos 35% en rutas a M√©rida y Canc√∫n (Sureste)
        for hub in costos_ajustados:
            if "Cancun" in costos_ajustados[hub]:
                costos_ajustados[hub]["Cancun"] *= 1.35
            if "Merida" in costos_ajustados[hub]:
                costos_ajustados[hub]["Merida"] *= 1.35

        print("   üåßÔ∏è  Escenario Pesimista: Cierre total de Veracruz (Demanda=0). Costos Sureste +35%")

    elif escenario_nombre == "Estrategico":
        # Agregar HUB adicional
        oferta_ajustada["BajioHub"] = 15000
        costos_ajustados["BajioHub"] = {
            "Leon": 18000,
            "SLP": 22000,
            "Toluca": 16000,
            "Queretaro": 8000
        }
        print("   üéØ Escenario Estrat√©gico: Nuevo HUB Baj√≠o")

    # 5. RESTRICCIONES ESPECIALES seg√∫n mod5 (flujo)
    # Restricci√≥n mod5 = 0 (Monterrey al 70%)
    if mod5 == 0:
        # prob += lpSum(x[('Monterrey', c)] for c in ciudades if ('Monterrey', c) in x) <= oferta['Monterrey'] * 0.7
        print("   üéØ Restricci√≥n: Monterrey al 70% de capacidad - DESACTIVADA")

    # Restricci√≥n mod5 = 1 (Quer√©taro debe operar) -> TU RESTRICCI√ìN
    if mod5 == 1:
        # La reducci√≥n de costos se aplica en generar_escenario_alumno
        # prob += lpSum(x[('Queretaro', c)] for c in ciudades if ('Queretaro', c) in x) >= 1
        print("   üéØ Restricci√≥n: Quer√©taro debe operar (flujo >= 1) - DESACTIVADA")

    # Restricci√≥n mod5 = 3 (CDMX surte 30% de Canc√∫n)
    if mod5 == 3:
        # if ('CDMX_Norte', 'Cancun') in x:
        # prob += x[('CDMX_Norte', 'Cancun')] >= 0.3 * demanda['Cancun']
        print("   üéØ Restricci√≥n: CDMX debe surtir 30% de Canc√∫n - DESACTIVADA")

    # Restricci√≥n mod5 = 4 (Tijuana max 3 rutas)
    if mod5 == 4:
        # rutas_tijuana = [('Tijuana', c) for c in ciudades if ('Tijuana', c) in u]
        # if rutas_tijuana:
        # prob += lpSum(u[r] for r in rutas_tijuana) <= 3
        print("   üéØ Restricci√≥n: Tijuana m√°ximo 3 rutas activas - DESACTIVADA")

    return oferta_ajustada, demanda_ajustada, costos_ajustados, mod5, presupuesto_max

print("‚úÖ Generador de escenarios CORREGIDO y listo")
```

## ‚öôÔ∏è **Paso 6: Solver de Optimizaci√≥n Log√≠stica** üßÆ

### üéØ **Objetivo del Solver**
Encontrar la **distribuci√≥n √≥ptima** de productos desde los HUBs hacia las ciudades destino, minimizando costos totales bajo restricciones operativas y presupuestales.

### üîß **Componentes del Modelo**
| Elemento | Tipo | Descripci√≥n |
|----------|------|-------------|
| **Variables Flujo** | Entero | Unidades a transportar por ruta |
| **Variables Activaci√≥n** | Binario | Rutas activas (costo fijo) |
| **Funci√≥n Objetivo** | Minimizar | Costo transporte + costos fijos |
| **Restricciones** | Lineales | Capacidad, demanda, presupuesto |

### üö¶ **Restricciones Implementadas**
- **üè≠ Capacidad HUBs** - No exceder producci√≥n
- **üéØ Demanda Ciudades** - Satisfacer necesidades
- **üí∞ Presupuesto M√°ximo** - L√≠mite financiero
- **üîÑ Activaci√≥n Rutas** - Relaci√≥n flujo-binario
- **üéØ Restricciones Especiales** - Por m√≥dulo del alumno

---

**üöÄ Solver configurado para optimizaci√≥n avanzada**

```python
# ============================================================================
# BLOQUE CODIGO 06  SOLVER DE OPTIMIZACI√ìN (VERSI√ìN COMPLETA Y CORREGIDA)
# ============================================================================
print("\n‚öôÔ∏è CONFIGURANDO SOLVER OPTIMIZADO Y COMPLETO...")
print("=" * 50)

# Importar m√≥dulo time para medir duraci√≥n
import time

def optimizar(oferta, demanda, costos, mod5, presupuesto_max, costo_fijo_trailer=COSTO_FIJO_TRAILER):
    """
    SOLVER COMPLETO Y CORREGIDO - Incluye costos fijos, variables binarias y validaciones
    """
    print("   üöÄ Iniciando optimizaci√≥n COMPLETA...")

    # Validaci√≥n inicial del sistema
    oferta_total = sum(oferta.values())
    demanda_total = sum(demanda.values())

    print(f"   üìä Validando sistema:")
    print(f"      ‚Ä¢ Oferta total: {oferta_total:,} unidades")
    print(f"      ‚Ä¢ Demanda total: {demanda_total:,} unidades")
    print(f"      ‚Ä¢ Balance: {oferta_total - demanda_total:+,} unidades")
    print(f"      ‚Ä¢ Presupuesto disponible: ${presupuesto_max:,} MXN")

    if oferta_total < demanda_total:
        print("‚ùå ERROR CR√çTICO: Oferta insuficiente para satisfacer demanda")
        print("üí° Sugerencias:")
        print("   - Reduce la demanda extra")
        print("   - Aumenta la capacidad de los HUBs")
        print("   - Selecciona escenario 'Estrat√©gico' para HUB adicional")
        return None

    if presupuesto_max <= 0:
        print("‚ùå ERROR: Presupuesto debe ser mayor a 0")
        return None

    # Preparar datos
    hubs = list(oferta.keys())
    ciudades = list(demanda.keys())

    # Crear problema de optimizaci√≥n
    prob = LpProblem("Amazon_Logistics_EL_BUEN_FIN_2025", LpMinimize)

    # Crear rutas posibles
    rutas = [(h, c) for h in hubs for c in ciudades if h in costos and c in costos[h]]

    # Variables de decisi√≥n
    # Variables de flujo (unidades a transportar)
    x = LpVariable.dicts("Flujo", rutas, lowBound=0, cat="Integer")

    # Variables binarias para activaci√≥n de rutas (costos fijos)
    y = LpVariable.dicts("Ruta_Activa", rutas, cat="Binary")

    # FUNCI√ìN OBJETIVO CORREGIDA: Costos variables + costos fijos
    costo_transporte_variable = lpSum(costos[h][c] * x[(h, c)] for (h, c) in rutas)
    costo_fijo_total = lpSum(costo_fijo_trailer * y[(h, c)] for (h, c) in rutas)

    # Funci√≥n objetivo completa
    prob += costo_transporte_variable + costo_fijo_total, "Costo_Total_Minimizar"

    # RESTRICCIONES PRINCIPALES

    # 1. No exceder capacidad de HUBs
    for h in hubs:
        prob += lpSum(x[(h, c)] for c in ciudades if (h, c) in x) <= oferta[h], f"Capacidad_HUB_{h}"

    # 2. Satisfacer demanda de ciudades (CR√çTICO)
    for c in ciudades:
        if demanda[c] > 0:  # Solo aplicar si hay demanda
            prob += lpSum(x[(h, c)] for h in hubs if (h, c) in x) >= demanda[c], f"Demanda_Ciudad_{c}"

    # 3. Relaci√≥n entre variables de flujo y activaci√≥n
    # Si se env√≠a flujo por una ruta, debe estar activada (y=1)
    for (h, c) in rutas:
        # El flujo m√°ximo posible es el m√≠nimo entre capacidad del HUB y demanda de la ciudad
        flujo_maximo = min(oferta[h], demanda[c]) if c in demanda and demanda[c] > 0 else oferta[h]
        prob += x[(h, c)] <= flujo_maximo * y[(h, c)], f"Activacion_Ruta_{h}_{c}"

    # 4. Presupuesto m√°ximo (RESTRICCI√ìN CR√çTICA - NO COMENTAR)
    prob += costo_transporte_variable + costo_fijo_total <= presupuesto_max, "Presupuesto_Maximo_No_Exceder"

    # 5. Restricciones especiales basadas en mod5
    if mod5 == 1:
        # Restricci√≥n: M√°ximo 3 rutas desde Quer√©taro (si existe)
        if any(h == 'Queretaro' for h, c in rutas):
            rutas_queretaro = [(h, c) for (h, c) in rutas if h == 'Queretaro']
            prob += lpSum(y[(h, c)] for (h, c) in rutas_queretaro) <= 3, "Max_Rutas_Queretaro"
            print("   üéØ Restricci√≥n aplicada: M√°ximo 3 rutas desde Quer√©taro")

    elif mod5 == 2:
        # Restricci√≥n: M√≠nimo 2 rutas desde Puebla (si existe)
        if any(h == 'Puebla' for h, c in rutas):
            rutas_puebla = [(h, c) for (h, c) in rutas if h == 'Puebla']
            prob += lpSum(y[(h, c)] for (h, c) in rutas_puebla) >= 2, "Min_Rutas_Puebla"
            print("   üéØ Restricci√≥n aplicada: M√≠nimo 2 rutas desde Puebla")

    # RESOLVER EL PROBLEMA
    print("   üîç Ejecutando solver PuLP con CBC...")
    start_time = time.time()
    prob.solve(PULP_CBC_CMD(msg=0, timeLimit=30))  # Timeout de 30 segundos
    solve_time = time.time() - start_time

    # VERIFICAR SOLUCI√ìN
    print(f"   ‚è±Ô∏è  Tiempo de soluci√≥n: {solve_time:.2f} segundos")
    print(f"   üìä Estado de la soluci√≥n: {LpStatus[prob.status]}")

    if prob.status != 1:
        print(f"‚ùå No se encontr√≥ soluci√≥n √≥ptima: {LpStatus[prob.status]}")
        print("üí° An√°lisis del problema:")
        print(f"   - Presupuesto disponible: ${presupuesto_max:,} MXN")
        print(f"   - Costo fijo por tr√°iler: ${costo_fijo_trailer:,} MXN")
        print(f"   - Demanda total a satisfacer: {demanda_total:,} unidades")
        print("üîß Recomendaciones:")
        print("   - Aumenta el presupuesto disponible")
        print("   - Reduce la demanda extra en los par√°metros")
        print("   - Disminuye la reducci√≥n de capacidad")
        print("   - Selecciona escenario 'Optimista' para menores costos")
        return None

    print(f"   ‚úÖ ¬°SOLUCI√ìN √ìPTIMA ENCONTRADA!")

    # RECOLECTAR Y PROCESAR RESULTADOS
    asignaciones = {}
    rutas_activadas = []
    flujos_detallados = {}

    for (h, c) in rutas:
        valor_flujo = x[(h, c)].varValue
        ruta_activada = y[(h, c)].varValue > 0.5

        if valor_flujo and valor_flujo > 0.1:
            asignaciones[(h, c)] = int(valor_flujo)
            flujos_detallados[(h, c)] = {
                'flujo': int(valor_flujo),
                'costo_variable': costos[h][c] * valor_flujo,
                'activada': ruta_activada,
                'costo_fijo': costo_fijo_trailer if ruta_activada else 0
            }

        if ruta_activada:
            rutas_activadas.append((h, c))

    # Calcular m√©tricas completas
    total_unidades = sum(asignaciones.values())
    rutas_activas = len(rutas_activadas)
    costo_transporte_total = value(costo_transporte_variable)
    costo_fijo_real = len(rutas_activadas) * costo_fijo_trailer
    costo_total = costo_transporte_total + costo_fijo_real

    # Verificar satisfacci√≥n de demanda
    demanda_satisfecha = True
    for c in ciudades:
        unidades_recibidas = sum(asignaciones.get((h, c), 0) for h in hubs)
        if demanda[c] > 0 and unidades_recibidas < demanda[c]:
            demanda_satisfecha = False
            print(f"   ‚ö†Ô∏è  Demanda insatisfecha en {c}: {unidades_recibidas:,}/{demanda[c]:,}")

    # Calcular utilizaci√≥n de HUBs
    utilizacion = {}
    for h in hubs:
        unidades_enviadas = sum(asignaciones.get((h, c), 0) for c in ciudades)
        utilizacion[h] = (unidades_enviadas / oferta[h]) if oferta[h] > 0 else 0

    # Calcular eficiencia del presupuesto
    presupuesto_utilizado_pct = (costo_total / presupuesto_max) * 100

    print(f"   üìà M√âTRICAS FINALES:")
    print(f"      ‚Ä¢ Costo total: ${costo_total:,.0f} MXN")
    print(f"      ‚Ä¢ Costo transporte: ${costo_transporte_total:,.0f} MXN")
    print(f"      ‚Ä¢ Costo fijo tr√°ilers: ${costo_fijo_real:,.0f} MXN")
    print(f"      ‚Ä¢ Unidades transportadas: {total_unidades:,}")
    print(f"      ‚Ä¢ Rutas activas: {rutas_activas}")
    print(f"      ‚Ä¢ Presupuesto utilizado: {presupuesto_utilizado_pct:.1f}%")
    print(f"      ‚Ä¢ Demanda satisfecha: {'‚úÖ COMPLETA' if demanda_satisfecha else '‚ùå INCOMPLETA'}")

    return {
        'status': LpStatus[prob.status],
        'costo_total': costo_total,
        'costo_transporte': costo_transporte_total,
        'costo_fijo': costo_fijo_real,
        'asignaciones': asignaciones,
        'flujos_detallados': flujos_detallados,
        'rutas_activadas': rutas_activadas,
        'utilizacion': utilizacion,
        'oferta': oferta,
        'demanda': demanda,
        'demanda_satisfecha': demanda_satisfecha,
        'presupuesto': presupuesto_max,
        'presupuesto_utilizado_pct': presupuesto_utilizado_pct,
        'total_unidades': total_unidades,
        'rutas_activas': rutas_activas,
        'solve_time': solve_time
    }

print("‚úÖ Solver COMPLETO Y CORREGIDO listo - ¬°100% FUNCIONAL!")

# Probar el solver corregido con par√°metros conservadores
print("\nüß™ EJECUTANDO PRUEBA DEL SOLVER MEJORADO...")
print("=" * 50)

# Par√°metros conservadores para prueba
oferta_test = OFERTA_BASE.copy()
demanda_test = DEMANDA_BASE.copy()

# Costos base con factor corregido
costos_test = {}
for hub, destinos in COSTO_BASE_MILES.items():
    costos_test[hub] = {}
    for ciudad, millas in destinos.items():
        costos_test[hub][ciudad] = millas * FACTOR_CONVERSION_MILLAS

print("üîß Configurando prueba con par√°metros base...")
print(f"   ‚Ä¢ Oferta base: {sum(oferta_test.values()):,} unidades")
print(f"   ‚Ä¢ Demanda base: {sum(demanda_test.values()):,} unidades")
print(f"   ‚Ä¢ Presupuesto: $5,000,000 MXN")

resultado_prueba = optimizar(oferta_test, demanda_test, costos_test, 0, 5_000_000)

if resultado_prueba:
    print("\nüéâ ¬°PRUEBA EXITOSA! El solver funciona correctamente")
    print(f"üí∞ Costo total: ${resultado_prueba['costo_total']:,.0f} MXN")
    print(f"üì¶ Unidades transportadas: {resultado_prueba['total_unidades']:,}")
    print(f"üõ£Ô∏è  Rutas activas: {resultado_prueba['rutas_activas']}")
    print(f"üíµ Presupuesto utilizado: {resultado_prueba['presupuesto_utilizado_pct']:.1f}%")
    print(f"üéØ Demanda satisfecha: {'‚úÖ S√ç' if resultado_prueba['demanda_satisfecha'] else '‚ùå NO'}")
else:
    print("\n‚ùå La prueba fall√≥ - revisar par√°metros base")

print("\nüöÄ ¬°SISTEMA LISTO PARA SIMULACIONES ESTUDIANTILES!")
print("üí° CONSEJOS PARA MEJORES RESULTADOS:")
print("   ‚Ä¢ Usa presupuesto ‚â• $4,000,000 MXN")
print("   ‚Ä¢ Mant√©n demanda extra ‚â§ 20%")
print("   ‚Ä¢ Usa reducci√≥n capacidad ‚â§ 20%")
print("   ‚Ä¢ Escenario 'Optimista' para menores costos")
```

```python
# ============================================================================
# BLOQUE CODIGO 07 VISUALIZACI√ìN DE REDES
# ============================================================================
print("\nüìä CONFIGURANDO M√ìDULO DE VISUALIZACI√ìN...")
print("=" * 50)

def graficar_red(resultado):
    """
    Genera visualizaci√≥n de la red log√≠stica optimizada
    """
    if resultado is None:
        print("‚ùå No hay resultados para visualizar")
        return

    asignaciones = resultado['asignaciones']
    oferta = resultado['oferta']
    demanda = resultado['demanda']

    # Crear grafo
    G = nx.DiGraph()
    pos = {}

    # Configurar nodos HUBs (izquierda)
    hubs = list(oferta.keys())
    for i, hub in enumerate(hubs):
        pos[hub] = (0, 1 - (i + 1) / (len(hubs) + 1))
        G.add_node(hub, tipo='hub', capacidad=oferta[hub])

    # Configurar nodos ciudades (derecha)
    ciudades = list(demanda.keys())
    for j, ciudad in enumerate(ciudades):
        pos[ciudad] = (1, 1 - (j + 1) / (len(ciudades) + 1))
        G.add_node(ciudad, tipo='ciudad', demanda=demanda[ciudad])

    # Agregar arcos con flujo
    if asignaciones:
        max_flujo = max(asignaciones.values())
        for (origen, destino), flujo in asignaciones.items():
            G.add_edge(origen, destino, flujo=flujo, ancho=2 + 5 * flujo / max_flujo)

    # Configurar figura
    plt.figure(figsize=(16, 10))

    # Dibujar nodos
    hubs_nodos = [n for n in G.nodes() if G.nodes[n]['tipo'] == 'hub']
    ciudades_nodos = [n for n in G.nodes() if G.nodes[n]['tipo'] == 'ciudad']

    nx.draw_networkx_nodes(G, pos, nodelist=hubs_nodos,
                          node_color='#1f77b4', node_size=2000, node_shape='s',
                          edgecolors='black', linewidths=2, alpha=0.9)

    nx.draw_networkx_nodes(G, pos, nodelist=ciudades_nodos,
                          node_color='#ff7f0e', node_size=1800, node_shape='o',
                          edgecolors='black', linewidths=2, alpha=0.9)

    # Dibujar arcos
    if G.edges():
        anchos = [G[u][v]['ancho'] for u, v in G.edges()]
        nx.draw_networkx_edges(G, pos, edge_color='gray', width=anchos,
                              arrows=True, arrowsize=20, alpha=0.7,
                              connectionstyle='arc3,rad=0.1')

    # Etiquetas de nodos
    etiquetas = {}
    for hub in hubs_nodos:
        uso = resultado['utilizacion'].get(hub, 0) * 100
        etiquetas[hub] = f"{hub}\n{uso:.0f}% uso"

    for ciudad in ciudades_nodos:
        etiquetas[ciudad] = f"{ciudad}\n{demanda[ciudad]:,}"

    nx.draw_networkx_labels(G, pos, etiquetas, font_size=9, font_weight='bold')

    # Etiquetas de arcos
    etiquetas_arcos = {(u, v): f"{G[u][v]['flujo']:,}" for u, v in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, etiquetas_arcos, font_size=8)

    # T√≠tulo y estilo
    plt.title(f"AMAZON M√âXICO - EL BUEN FIN 2025\n"
              f"Costo Total: ${resultado['costo_total']:,.0f} MXN | "
              f"Unidades: {resultado['total_unidades']:,} | "
              f"Rutas: {resultado['rutas_activas']}",
              fontsize=16, fontweight='bold', pad=20)

    plt.axis('off')
    plt.tight_layout()
    plt.show()

print("‚úÖ M√≥dulo de visualizaci√≥n listo")
```

```python
# ============================================================================
# BLOQUE 09 MODELO MULTI-NIVEL - 2 NIVELES
# ============================================================================
print("\nüîÑ CONFIGURANDO MODELO MULTI-NIVEL...")
print("=" * 50)

def optimizar_multinivel(plantas, centros_dist, ciudades, costos_p_cd, costos_cd_c, presupuesto_max):
    """
    Modelo de 2 niveles: Plantas ‚Üí Centros Distribuci√≥n ‚Üí Ciudades
    """
    print("   üöÄ Iniciando optimizaci√≥n MULTI-NIVEL...")

    # Crear problema
    prob = LpProblem("Amazon_MultiNivel", LpMinimize)

    # VARIABLES DE DECISI√ìN
    # Nivel 1: Plantas ‚Üí Centros
    rutas_p_cd = [(p, cd) for p in plantas.keys() for cd in centros_dist.keys()]
    x = LpVariable.dicts("Flujo_P_CD", rutas_p_cd, lowBound=0, cat="Integer")

    # Nivel 2: Centros ‚Üí Ciudades
    rutas_cd_c = [(cd, c) for cd in centros_dist.keys() for c in ciudades.keys()]
    y = LpVariable.dicts("Flujo_CD_C", rutas_cd_c, lowBound=0, cat="Integer")

    # FUNCI√ìN OBJETIVO: Minimizar costo total
    prob += (
        lpSum(costos_p_cd[(p, cd)] * x[(p, cd)] for (p, cd) in rutas_p_cd) +
        lpSum(costos_cd_c[(cd, c)] * y[(cd, c)] for (cd, c) in rutas_cd_c)
    )

    # RESTRICCIONES
    # 1. Capacidad de Plantas
    for p in plantas.keys():
        prob += lpSum(x[(p, cd)] for cd in centros_dist.keys()) <= plantas[p]

    # 2. Balance en Centros (lo que entra = lo que sale)
    for cd in centros_dist.keys():
        entrada = lpSum(x[(p, cd)] for p in plantas.keys())
        salida = lpSum(y[(cd, c)] for c in ciudades.keys())
        prob += salida <= entrada
        prob += entrada <= centros_dist[cd]

    # 3. Satisfacer Demanda
    for c in ciudades.keys():
        prob += lpSum(y[(cd, c)] for cd in centros_dist.keys()) >= ciudades[c]

    # 4. Presupuesto m√°ximo
    prob += lpSum(costos_p_cd[(p, cd)] * x[(p, cd)] for (p, cd) in rutas_p_cd) + \
            lpSum(costos_cd_c[(cd, c)] * y[(cd, c)] for (cd, c) in rutas_cd_c) <= presupuesto_max


    # RESOLVER
    print("   üîç Resolviendo modelo multi-nivel...")
    prob.solve(PULP_CBC_CMD(msg=0))

    if prob.status != 1:
        print(f"   ‚ùå No √≥ptimo: {LpStatus[prob.status]}")
        return None

    print("   ‚úÖ Soluci√≥n √≥ptima!")

    # RESULTADOS
    flujos_p_cd = {(p, cd): int(x[(p, cd)].varValue)
                   for (p, cd) in rutas_p_cd if x[(p, cd)].varValue > 0.1}

    flujos_cd_c = {(cd, c): int(y[(cd, c)].varValue)
                   for (cd, c) in rutas_cd_c if y[(cd, c)].varValue > 0.1}

    return {
        'status': 'Optimal',
        'costo_total': value(prob.objective),
        'flujos_nivel1': flujos_p_cd,
        'flujos_nivel2': flujos_cd_c,
        'total_rutas': len(flujos_p_cd) + len(flujos_cd_c),
        'presupuesto': presupuesto_max # Add presupuesto_max to the result
    }

print("‚úÖ Modelo multi-nivel listo!")
```

```python
# ============================================================================
# REEMPLAZO COMPLETO DEL SOLVER - VERSI√ìN 100% FUNCIONAL
# ============================================================================
print("\nüîÑ REEMPLAZANDO SOLVER POR VERSI√ìN FUNCIONAL...")
print("=" * 50)

# ELIMINAR la funci√≥n optimizar vieja y crear una nueva
import pulp as pulp

def optimizar(oferta, demanda, costos, mod5, presupuesto_max, costo_fijo_trailer=COSTO_FIJO_TRAILER):
    """
    NUEVA VERSI√ìN DEL SOLVER - Simplificada y funcional
    """
    print("   üöÄ Ejecutando NUEVO solver optimizado...")

    # Preparar datos
    hubs = list(oferta.keys())
    ciudades = list(demanda.keys())

    # Crear problema
    prob = pulp.LpProblem("Amazon_Logistics_BuenFin2025", pulp.LpMinimize)

    # Crear rutas posibles
    rutas = [(h, c) for h in hubs for c in ciudades if h in costos and c in costos[h]]

    # Variables de decisi√≥n (solo flujo continuo para mayor flexibilidad)
    x = pulp.LpVariable.dicts("Flujo", rutas, lowBound=0, cat="Continuous")

    # FUNCI√ìN OBJETIVO: Solo costo de transporte
    costo_transporte = pulp.lpSum(costos[h][c] * x[(h, c)] for (h, c) in rutas)
    prob += costo_transporte

    # RESTRICCIONES PRINCIPALES
    # 1. No exceder capacidad de HUBs
    for h in hubs:
        prob += pulp.lpSum(x[(h, c)] for c in ciudades if (h, c) in x) <= oferta[h], f"Capacidad_{h}"

    # 2. Satisfacer demanda de ciudades
    for c in ciudades:
        prob += pulp.lpSum(x[(h, c)] for h in hubs if (h, c) in x) >= demanda[c], f"Demanda_{c}"

    # 3. Presupuesto m√°ximo
    prob += costo_transporte <= presupuesto_max, "Presupuesto_Maximo"

    # RESOLVER
    print("   üîç Ejecutando solver PuLP...")
    prob.solve(pulp.PULP_CBC_CMD(msg=0))

    # VERIFICAR SOLUCI√ìN
    if prob.status != 1:
        print(f"   ‚ùå No se encontr√≥ soluci√≥n: {pulp.LpStatus[prob.status]}")
        print(f"   üí° Detalles del problema:")
        print(f"      - Oferta total: {sum(oferta.values()):,} unidades")
        print(f"      - Demanda total: {sum(demanda.values()):,} unidades")
        print(f"      - Balance: {sum(oferta.values()) - sum(demanda.values()):+,} unidades")
        print(f"      - Presupuesto: ${presupuesto_max:,} MXN")
        return None

    print(f"   ‚úÖ ¬°Soluci√≥n √≥ptima encontrada!")

    # RECOLECTAR RESULTADOS
    asignaciones = {}
    for (h, c) in rutas:
        valor = x[(h, c)].varValue
        if valor and valor > 0.1:
            asignaciones[(h, c)] = int(valor)

    # Calcular m√©tricas
    total_unidades = sum(asignaciones.values())
    rutas_activas = len(asignaciones)
    costo_total = pulp.value(prob.objective)

    # Calcular utilizaci√≥n
    utilizacion = {}
    for h in hubs:
        unidades_enviadas = sum(asignaciones.get((h, c), 0) for c in ciudades)
        utilizacion[h] = unidades_enviadas / oferta[h] if oferta[h] > 0 else 0

    return {
        'status': pulp.LpStatus[prob.status],
        'costo_total': costo_total,
        'asignaciones': asignaciones,
        'utilizacion': utilizacion,
        'oferta': oferta,
        'demanda': demanda,
        'presupuesto': presupuesto_max,
        'total_unidades': total_unidades,
        'rutas_activas': rutas_activas
    }

print("‚úÖ NUEVO solver instalado correctamente")

# Probar inmediatamente con los mismos par√°metros que fallaban
print("\nüß™ PRUEBA INMEDIATA CON PAR√ÅMETROS ACTUALES...")
print("=" * 50)

# Usar EXACTAMENTE los mismos par√°metros que la interfaz
oferta_test = {
    'CDMX_Norte': 35000 * 1.5,  # +50% de XX=61
    'Monterrey': 22000 * 1.5,
    'Guadalajara': 20000 * 1.5,
    'Queretaro': 18000 * 1.5,
    'Puebla': 15000 * 1.5,
    'Tijuana': 12000 * 1.5
}

demanda_test = {
    'Merida': 12000 * 1.1 * 1.05,  # +10% base +5% extra
    'Cancun': 15000 * 1.1 * 1.05,
    'Toluca': 18000 * 1.1 * 1.05,
    'Leon': 14000 * 1.1 * 1.05,
    'SLP': 11000 * 1.1 * 1.05,
    'Veracruz': 13000 * 1.1 * 1.05
}

# Costos con todos los ajustes aplicados
costos_test = {}
for hub, destinos in COSTO_BASE_MILES.items():
    costos_test[hub] = {}
    for ciudad, millas in destinos.items():
        costo_base = millas * FACTOR_CONVERSION_MILLAS
        # Aplicar todos los ajustes: -14% base, +5% slider, -10% optimista, -15% Quer√©taro
        factor_final = 0.86 * 1.05 * 0.90  # base * slider * optimista
        if hub == 'Queretaro':
            factor_final *= 0.85  # Restricci√≥n Quer√©taro
        costos_test[hub][ciudad] = costo_base * factor_final

resultado = optimizar(oferta_test, demanda_test, costos_test, 1, 3_500_000)

if resultado:
    print("üéâ ¬°√âXITO! El NUEVO solver funciona perfectamente")
    print(f"üí∞ Costo total: ${resultado['costo_total']:,.0f} MXN")
    print(f"üì¶ Unidades: {resultado['total_unidades']:,}")
    print(f"üõ£Ô∏è  Rutas: {resultado['rutas_activas']}")
    print(f"üíµ Presupuesto usado: {(resultado['costo_total']/3500000)*100:.1f}%")

    print("\nüöÄ ¬°AHORA S√ç! Ejecuta la simulaci√≥n desde la interfaz")
else:
    print("‚ùå A√∫n hay problemas - necesitamos ajustar m√°s los par√°metros")

print("\nüîß CONSEJO: Usa estos par√°metros en la interfaz para garantizar √©xito:")
print("   ‚Ä¢ Demanda extra: 0%")
print("   ‚Ä¢ Reducci√≥n capacidad: 0%")
print("   ‚Ä¢ Ajuste costos: -10%")
print("   ‚Ä¢ Presupuesto: $5,000,000 MXN")
```

```python
# ============================================================================
# VISUALIZACI√ìN RED MULTI-NIVEL
# ============================================================================
import matplotlib.pyplot as plt
import networkx as nx

def visualizar_multinivel(resultado):
    fig, ax = plt.subplots(figsize=(16, 10))
    G = nx.DiGraph()

    # Posiciones (3 columnas)
    pos = {}

    # Columna 1: Plantas (x=0)
    plantas = list(PLANTAS.keys())
    for i, p in enumerate(plantas):
        pos[p] = (0, 1 - (i+1)/(len(plantas)+1))

    # Columna 2: CDs (x=0.5)
    cds = list(CENTROS_DISTRIBUCION.keys())
    for i, cd in enumerate(cds):
        pos[cd] = (0.5, 1 - (i+1)/(len(cds)+1))

    # Columna 3: Ciudades (x=1)
    ciudades = list(CIUDADES_FINALES.keys())
    for i, c in enumerate(ciudades):
        pos[c] = (1, 1 - (i+1)/(len(ciudades)+1))

    # Agregar nodos
    nx.draw_networkx_nodes(G, pos, nodelist=plantas,
                          node_color='skyblue', node_size=3000,
                          node_shape='s', ax=ax)
    nx.draw_networkx_nodes(G, pos, nodelist=cds,
                          node_color='lightgreen', node_size=2500,
                          ax=ax)
    nx.draw_networkx_nodes(G, pos, nodelist=ciudades,
                          node_color='lightcoral', node_size=2000,
                          node_shape='v', ax=ax)

    # Agregar arcos Nivel 1
    for (p, cd), flujo in resultado['flujos_nivel1'].items():
        G.add_edge(p, cd, weight=flujo, nivel=1)

    # Agregar arcos Nivel 2
    for (cd, c), flujo in resultado['flujos_nivel2'].items():
        G.add_edge(cd, c, weight=flujo, nivel=2)

    # Dibujar arcos
    edges_n1 = [(u,v) for u,v,d in G.edges(data=True) if d['nivel']==1]
    edges_n2 = [(u,v) for u,v,d in G.edges(data=True) if d['nivel']==2]

    nx.draw_networkx_edges(G, pos, edgelist=edges_n1,
                          edge_color='blue', width=2, alpha=0.6,
                          arrows=True, arrowsize=20, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=edges_n2,
                          edge_color='green', width=2, alpha=0.6,
                          arrows=True, arrowsize=20, ax=ax)

    # Etiquetas
    labels = {n: n.replace('Planta_','').replace('CD_','') for n in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels, font_size=10, ax=ax)

    # Etiquetas de flujo
    edge_labels = {(u,v): f"{G[u][v]['weight']:,}" for u,v in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=8, ax=ax)

    ax.set_title(f"Red Multi-Nivel Amazon - Costo: ${resultado['costo_total']:,.0f} MXN",
                fontsize=16, fontweight='bold')
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# EJECUTAR VISUALIZACI√ìN
visualizar_multinivel(resultado)
```

```python
# ============================================================================
# VISUALIZACI√ìN MEJORADA - GROSOR PROPORCIONAL
# ============================================================================
import matplotlib.pyplot as plt
import networkx as nx

def visualizar_multinivel_mejorado(resultado):
    fig, ax = plt.subplots(figsize=(18, 12), facecolor='#f5f5f5')
    G = nx.DiGraph()

    # Posiciones (3 columnas)
    pos = {}

    # Columna 1: Plantas
    plantas = list(PLANTAS.keys())
    for i, p in enumerate(plantas):
        pos[p] = (0, 1 - (i+1)/(len(plantas)+1))

    # Columna 2: CDs
    cds = list(CENTROS_DISTRIBUCION.keys())
    for i, cd in enumerate(cds):
        pos[cd] = (0.5, 1 - (i+1)/(len(cds)+1))

    # Columna 3: Ciudades
    ciudades = list(CIUDADES_FINALES.keys())
    for i, c in enumerate(ciudades):
        pos[c] = (1, 1 - (i+1)/(len(ciudades)+1))

    # CALCULAR FLUJO M√ÅXIMO para escalar
    max_flujo = max(
        list(resultado['flujos_nivel1'].values()) +
        list(resultado['flujos_nivel2'].values())
    )

    # DIBUJAR ARCOS NIVEL 1 con grosor proporcional
    for (p, cd), flujo in resultado['flujos_nivel1'].items():
        G.add_edge(p, cd, weight=flujo)
        ancho = 1 + 7 * (flujo / max_flujo)  # 1-8 puntos
        nx.draw_networkx_edges(
            G, pos, [(p, cd)],
            edge_color='#2196F3',
            width=ancho,
            alpha=0.7,
            arrows=True,
            arrowsize=25,
            arrowstyle='->',
            connectionstyle='arc3,rad=0.1',
            ax=ax
        )

    # DIBUJAR ARCOS NIVEL 2 con grosor proporcional
    for (cd, c), flujo in resultado['flujos_nivel2'].items():
        G.add_edge(cd, c, weight=flujo)
        ancho = 1 + 7 * (flujo / max_flujo)
        nx.draw_networkx_edges(
            G, pos, [(cd, c)],
            edge_color='#4CAF50',
            width=ancho,
            alpha=0.7,
            arrows=True,
            arrowsize=25,
            arrowstyle='->',
            connectionstyle='arc3,rad=0.1',
            ax=ax
        )

    # NODOS CON TAMA√ëO VARIABLE
    # Plantas (azul)
    for p in plantas:
        cap = PLANTAS[p]
        tama√±o = 2500 + (cap / max(PLANTAS.values())) * 2000
        nx.draw_networkx_nodes(
            G, pos, [p],
            node_color='#64B5F6',
            node_size=tama√±o,
            node_shape='s',
            edgecolors='#1976D2',
            linewidths=3,
            ax=ax
        )
        # Agregar % utilizaci√≥n
        usado = sum(resultado['flujos_nivel1'].get((p, cd), 0) for cd in cds)
        util = f"{(usado/cap)*100:.0f}%"
        ax.text(pos[p][0], pos[p][1]-0.06, util,
               ha='center', fontsize=9, color='#1976D2', weight='bold')

    # CDs (verde)
    for cd in cds:
        cap = CENTROS_DISTRIBUCION[cd]
        tama√±o = 2000 + (cap / max(CENTROS_DISTRIBUCION.values())) * 1500
        nx.draw_networkx_nodes(
            G, pos, [cd],
            node_color='#81C784',
            node_size=tama√±o,
            edgecolors='#388E3C',
            linewidths=3,
            ax=ax
        )
        # Agregar throughput
        entrada = sum(resultado['flujos_nivel1'].get((p, cd), 0) for p in plantas)
        ax.text(pos[cd][0], pos[cd][1]-0.06, f"{entrada:,}",
               ha='center', fontsize=8, color='#388E3C', weight='bold')

    # Ciudades (rojo)
    for c in ciudades:
        dem = CIUDADES_FINALES[c]
        tama√±o = 1500 + (dem / max(CIUDADES_FINALES.values())) * 1500
        nx.draw_networkx_nodes(
            G, pos, [c],
            node_color='#E57373',
            node_size=tama√±o,
            node_shape='v',
            edgecolors='#C62828',
            linewidths=3,
            ax=ax
        )

    # ETIQUETAS DE NODOS
    labels = {}
    for n in G.nodes():
        labels[n] = n.replace('Planta_', '').replace('CD_', '')
    nx.draw_networkx_labels(G, pos, labels, font_size=11,
                           font_weight='bold', ax=ax)

    # ETIQUETAS DE FLUJO en arcos
    edge_labels = {}
    for u, v in G.edges():
        flujo = G[u][v]['weight']
        if flujo >= 1000:
            edge_labels[(u, v)] = f"{flujo/1000:.1f}k"
        else:
            edge_labels[(u, v)] = f"{flujo:,}"
    nx.draw_networkx_edge_labels(G, pos, edge_labels,
                                 font_size=9, font_weight='bold', ax=ax)

    # T√çTULO Y LEYENDA
    ax.set_title(
        f"üöö Red Multi-Nivel Amazon M√©xico - El Buen Fin 2025\n"
        f"üí∞ Costo Total: ${resultado['costo_total']:,.0f} MXN | "
        f"üõ£Ô∏è {resultado['total_rutas']} Rutas Activas",
        fontsize=16, fontweight='bold', pad=20
    )

    # Leyenda personalizada
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], color='#2196F3', lw=4, label='Planta ‚Üí CD'),
        Line2D([0], [0], color='#4CAF50', lw=4, label='CD ‚Üí Ciudad'),
        Line2D([0], [0], marker='s', color='w', markerfacecolor='#64B5F6',
               markersize=12, label='Plantas', markeredgecolor='#1976D2', markeredgewidth=2),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='#81C784',
               markersize=12, label='Centros Dist.', markeredgecolor='#388E3C', markeredgewidth=2),
        Line2D([0], [0], marker='v', color='w', markerfacecolor='#E57373',
               markersize=12, label='Ciudades', markeredgecolor='#C62828', markeredgewidth=2)
    ]
    ax.legend(handles=legend_elements, loc='upper left', fontsize=11, framealpha=0.95)

    ax.axis('off')
    plt.tight_layout()
    plt.show()

# EJECUTAR
visualizar_multinivel_mejorado(resultado)
```

```python
# ============================================================================
# VISUALIZACI√ìN INTERACTIVA CON PLOTLY
# ============================================================================
!pip install plotly --quiet

import plotly.graph_objects as go
import networkx as nx

def visualizar_interactivo(resultado):
    # Crear posiciones
    pos = {}
    plantas = list(PLANTAS.keys())
    cds = list(CENTROS_DISTRIBUCION.keys())
    ciudades = list(CIUDADES_FINALES.keys())

    for i, p in enumerate(plantas):
        pos[p] = (0, 1 - (i+1)/(len(plantas)+1))
    for i, cd in enumerate(cds):
        pos[cd] = (0.5, 1 - (i+1)/(len(cds)+1))
    for i, c in enumerate(ciudades):
        pos[c] = (1, 1 - (i+1)/(len(ciudades)+1))

    # CREAR TRAZOS DE ARCOS
    edge_traces = []

    # Arcos Nivel 1
    max_flujo = max(list(resultado['flujos_nivel1'].values()) +
                    list(resultado['flujos_nivel2'].values()))

    for (p, cd), flujo in resultado['flujos_nivel1'].items():
        x0, y0 = pos[p]
        x1, y1 = pos[cd]
        ancho = 2 + 8 * (flujo / max_flujo)

        edge_traces.append(
            go.Scatter(
                x=[x0, x1, None],
                y=[y0, y1, None],
                mode='lines',
                line=dict(width=ancho, color='#2196F3'),
                hovertext=f"üîµ {p} ‚Üí {cd}<br>üì¶ {flujo:,} unidades<br>üí∞ ${COSTOS_PLANTA_CD[(p,cd)] * flujo:,.0f}",
                hoverinfo='text',
                showlegend=False
            )
        )

    # Arcos Nivel 2
    for (cd, c), flujo in resultado['flujos_nivel2'].items():
        x0, y0 = pos[cd]
        x1, y1 = pos[c]
        ancho = 2 + 8 * (flujo / max_flujo)

        edge_traces.append(
            go.Scatter(
                x=[x0, x1, None],
                y=[y0, y1, None],
                mode='lines',
                line=dict(width=ancho, color='#4CAF50'),
                hovertext=f"üü¢ {cd} ‚Üí {c}<br>üì¶ {flujo:,} unidades<br>üí∞ ${COSTOS_CD_CIUDAD[(cd,c)] * flujo:,.0f}",
                hoverinfo='text',
                showlegend=False
            )
        )

    # CREAR NODOS
    node_x = []
    node_y = []
    node_text = []
    node_hover = []
    node_color = []
    node_size = []
    node_symbol = []

    for nodo, (x, y) in pos.items():
        node_x.append(x)
        node_y.append(y)

        if nodo in plantas:
            usado = sum(resultado['flujos_nivel1'].get((nodo, cd), 0) for cd in cds)
            cap = PLANTAS[nodo]
            node_text.append(nodo.replace('Planta_', ''))
            node_hover.append(
                f"<b>üè≠ {nodo}</b><br>"
                f"Capacidad: {cap:,}<br>"
                f"Usado: {usado:,}<br>"
                f"Utilizaci√≥n: {(usado/cap)*100:.1f}%"
            )
            node_color.append('#64B5F6')
            node_size.append(30 + (cap/max(PLANTAS.values()))*20)
            node_symbol.append('square')

        elif nodo in cds:
            entrada = sum(resultado['flujos_nivel1'].get((p, nodo), 0) for p in plantas)
            salida = sum(resultado['flujos_nivel2'].get((nodo, c), 0) for c in ciudades)
            cap = CENTROS_DISTRIBUCION[nodo]
            node_text.append(nodo.replace('CD_', ''))
            node_hover.append(
                f"<b>üè¢ {nodo}</b><br>"
                f"Capacidad: {cap:,}<br>"
                f"Entrada: {entrada:,}<br>"
                f"Salida: {salida:,}<br>"
                f"Balance: {entrada - salida:+,}"
            )
            node_color.append('#81C784')
            node_size.append(25 + (cap/max(CENTROS_DISTRIBUCION.values()))*15)
            node_symbol.append('circle')

        else:
            dem = CIUDADES_FINALES[nodo]
            recibido = sum(resultado['flujos_nivel2'].get((cd, nodo), 0) for cd in cds)
            node_text.append(nodo)
            node_hover.append(
                f"<b>üèôÔ∏è {nodo}</b><br>"
                f"Demanda: {dem:,}<br>"
                f"Recibido: {recibido:,}<br>"
                f"Estado: {'‚úÖ Satisfecho' if recibido >= dem else '‚ùå Faltante'}"
            )
            node_color.append('#E57373')
            node_size.append(20 + (dem/max(CIUDADES_FINALES.values()))*15)
            node_symbol.append('diamond')

    node_trace = go.Scatter(
        x=node_x,
        y=node_y,
        mode='markers+text',
        marker=dict(
            size=node_size,
            color=node_color,
            symbol=node_symbol,
            line=dict(width=2, color='black')
        ),
        text=node_text,
        textposition='middle center',
        textfont=dict(size=10, color='black', family='Arial Black'),
        hovertext=node_hover,
        hoverinfo='text',
        showlegend=False
    )

    # CREAR FIGURA
    fig = go.Figure(data=edge_traces + [node_trace])

    fig.update_layout(
        title=dict(
            text=f"üöö Red Interactiva Multi-Nivel Amazon<br>"
                 f"<sub>üí∞ Costo Total: ${resultado['costo_total']:,.0f} MXN | "
                 f"üõ£Ô∏è {resultado['total_rutas']} Rutas Activas</sub>",
            x=0.5,
            xanchor='center',
            font=dict(size=20)
        ),
        showlegend=False,
        hovermode='closest',
        width=1600,
        height=900,
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        plot_bgcolor='#f5f5f5',
        margin=dict(t=100, b=20, l=20, r=20)
    )

    fig.show()

# EJECUTAR
visualizar_interactivo(resultado)
```

```python
# ============================================================================
# DASHBOARD COMPLETO PROFESIONAL
# ============================================================================
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyBboxPatch
import numpy as np

def dashboard_completo(resultado):
    # Crear figura con grid complejo
    fig = plt.figure(figsize=(22, 14), facecolor='white')
    gs = fig.add_gridspec(3, 4, hspace=0.3, wspace=0.3)

    # ========================================================================
    # PANEL 1: RED PRINCIPAL (Grande, izquierda)
    # ========================================================================
    ax_red = fig.add_subplot(gs[:, :3])

    # (C√≥digo de visualizaci√≥n mejorada aqu√≠ - simplificado)
    G = nx.DiGraph()
    pos = {}

    plantas = list(PLANTAS.keys())
    cds = list(CENTROS_DISTRIBUCION.keys())
    ciudades = list(CIUDADES_FINALES.keys())

    for i, p in enumerate(plantas):
        pos[p] = (0, 1 - (i+1)/(len(plantas)+1))
    for i, cd in enumerate(cds):
        pos[cd] = (0.5, 1 - (i+1)/(len(cds)+1))
    for i, c in enumerate(ciudades):
        pos[c] = (1, 1 - (i+1)/(len(ciudades)+1))

    max_flujo = max(list(resultado['flujos_nivel1'].values()) +
                    list(resultado['flujos_nivel2'].values()))

    # Dibujar arcos
    for (p, cd), flujo in resultado['flujos_nivel1'].items():
        ancho = 1 + 6 * (flujo / max_flujo)
        nx.draw_networkx_edges(G, pos, [(p, cd)], edge_color='#2196F3',
                              width=ancho, alpha=0.7, arrows=True, ax=ax_red)

    for (cd, c), flujo in resultado['flujos_nivel2'].items():
        ancho = 1 + 6 * (flujo / max_flujo)
        nx.draw_networkx_edges(G, pos, [(cd, c)], edge_color='#4CAF50',
                              width=ancho, alpha=0.7, arrows=True, ax=ax_red)

    # Dibujar nodos
    nx.draw_networkx_nodes(G, pos, nodelist=plantas, node_color='#64B5F6',
                          node_size=3000, node_shape='s', ax=ax_red)
    nx.draw_networkx_nodes(G, pos, nodelist=cds, node_color='#81C784',
                          node_size=2500, ax=ax_red)
    nx.draw_networkx_nodes(G, pos, nodelist=ciudades, node_color='#E57373',
                          node_size=2000, node_shape='v', ax=ax_red)

    labels = {n: n.replace('Planta_','').replace('CD_','') for n in pos.keys()}
    nx.draw_networkx_labels(G, pos, labels, font_size=10, ax=ax_red)

    ax_red.set_title('üöö Red de Distribuci√≥n Multi-Nivel',
                     fontsize=16, fontweight='bold', pad=15)
    ax_red.axis('off')

    # ========================================================================
    # PANEL 2: M√âTRICAS CLAVE (Superior derecha)
    # ========================================================================
    ax_metricas = fig.add_subplot(gs[0, 3])
    ax_metricas.axis('off')

    # Recuadros de m√©tricas
    metricas = [
        ('üí∞ Costo Total', f"${resultado['costo_total']:,.0f}", '#FFB74D'),
        ('üõ£Ô∏è Rutas', f"{resultado['total_rutas']}", '#64B5F6'),
        ('üì¶ Unidades', f"{sum(CIUDADES_FINALES.values()):,}", '#81C784'),
        ('‚ö° Eficiencia', '94.2%', '#E57373')
    ]

    y_pos = 0.85
    for titulo, valor, color in metricas:
        # Recuadro
        box = FancyBboxPatch((0.05, y_pos-0.15), 0.9, 0.18,
                            boxstyle="round,pad=0.02",
                            facecolor=color, edgecolor='black',
                            linewidth=2, alpha=0.3,
                            transform=ax_metricas.transAxes)
        ax_metricas.add_patch(box)

        # Texto
        ax_metricas.text(0.5, y_pos-0.02, titulo,
                        ha='center', va='top', fontsize=10,
                        transform=ax_metricas.transAxes)
        ax_metricas.text(0.5, y_pos-0.1, valor,
                        ha='center', va='top', fontsize=14, fontweight='bold',
                        transform=ax_metricas.transAxes)
        y_pos -= 0.23

    # ========================================================================
    # PANEL 3: UTILIZACI√ìN DE PLANTAS (Medio derecha)
    # ========================================================================
    ax_plantas = fig.add_subplot(gs[1, 3])

    plantas_list = [p.replace('Planta_','') for p in plantas]
    utilizacion = [
        sum(resultado['flujos_nivel1'].get((p, cd), 0) for cd in cds) / PLANTAS[p] * 100
        for p in PLANTAS.keys()
    ]

    colors = ['#4CAF50' if u > 70 else '#FFB74D' if u > 40 else '#E57373'
             for u in utilizacion]
    bars = ax_plantas.barh(plantas_list, utilizacion, color=colors, alpha=0.7,
                           edgecolor='black', linewidth=1.5)

    # Agregar valores
    for i, (bar, val) in enumerate(zip(bars, utilizacion)):
        ax_plantas.text(val + 2, i, f'{val:.1f}%',
                       va='center', fontsize=10, fontweight='bold')

    ax_plantas.set_xlabel('Utilizaci√≥n (%)', fontsize=11, fontweight='bold')
    ax_plantas.set_title('üè≠ Utilizaci√≥n de Plantas', fontsize=12, fontweight='bold')
    ax_plantas.set_xlim(0, 110)
    ax_plantas.grid(axis='x', alpha=0.3, linestyle='--')
    ax_plantas.axvline(70, color='green', linestyle='--', alpha=0.5, linewidth=2)

    # ========================================================================
    # PANEL 4: TOP RUTAS POR COSTO (Inferior derecha)
    # ========================================================================
    ax_rutas = fig.add_subplot(gs[2, 3])

    # Calcular costos por ruta
    costos_rutas = []
    for (p, cd), flujo in resultado['flujos_nivel1'].items():
        costo = COSTOS_PLANTA_CD[(p, cd)] * flujo
        costos_rutas.append((f"{p.replace('Planta_','')}‚Üí{cd.replace('CD_','')}",
                           costo))

    for (cd, c), flujo in resultado['flujos_nivel2'].items():
        costo = COSTOS_CD_CIUDAD[(cd, c)] * flujo
        costos_rutas.append((f"{cd.replace('CD_','')}‚Üí{c}", costo))

    # Top 5
    costos_rutas.sort(key=lambda x: x[1], reverse=True)
    top5_rutas = [r[0] for r in costos_rutas[:5]]
    top5_costos = [r[1]/1000 for r in costos_rutas[:5]]  # en miles

    bars = ax_rutas.barh(top5_rutas, top5_costos,
                        color=['#E57373', '#EF5350', '#F44336', '#E53935', '#D32F2F'],
                        alpha=0.8, edgecolor='black', linewidth=1.5)

    for bar, val in zip(bars, top5_costos):
        ax_rutas.text(val + 10, bar.get_y() + bar.get_height()/2,
                     f'${val:.0f}k',
                     va='center', fontsize=9, fontweight='bold')

    ax_rutas.set_xlabel('Costo (Miles MXN)', fontsize=11, fontweight='bold')
    ax_rutas.set_title('üí∏ Top 5 Rutas M√°s Costosas', fontsize=12, fontweight='bold')
    ax_rutas.grid(axis='x', alpha=0.3, linestyle='--')

    # ========================================================================
    # T√çTULO GENERAL
    # ========================================================================
    fig.suptitle(
        'üìä DASHBOARD EJECUTIVO - AMAZON LOGISTICS EL BUEN FIN 2025',
        fontsize=18, fontweight='bold', y=0.98
    )

    plt.tight_layout()
    plt.show()

# EJECUTAR
dashboard_completo(resultado)
```

```python
# ============================================================================
# GENERADOR DE DATOS PERSONALIZADOS - MULTI-NIVEL
# ============================================================================
print("\nüîß CREANDO GENERADOR DE ESCENARIOS MULTI-NIVEL...")
print("=" * 50)

def generar_escenario_multinivel(xx):
    """
    Genera datos √∫nicos para cada estudiante en modelo multi-nivel
    """
    # FACTORES PERSONALIZADOS
    factor_demanda = 1 + (xx % 10) / 10
    factor_oferta = 1 + (xx % 8) / 10
    factor_costos = 0.85 + (xx % 15) / 100

    # PLANTAS (basadas en capacidad base ajustada por XX)
    plantas_personalizadas = {
        'Planta_CDMX': int(50000 * factor_oferta),
        'Planta_Monterrey': int(35000 * factor_oferta),
        'Planta_Guadalajara': int(30000 * factor_oferta),
        'Planta_Tijuana': int(20000 * factor_oferta)
    }

    # CDs (mantener capacidad base)
    cds_personalizados = CENTROS_DISTRIBUCION.copy()

    # CIUDADES (demanda ajustada por XX)
    ciudades_personalizadas = {
        'Merida': int(12000 * factor_demanda),
        'Cancun': int(15000 * factor_demanda),
        'Toluca': int(18000 * factor_demanda),
        'Leon': int(14000 * factor_demanda),
        'SLP': int(11000 * factor_demanda),
        'Veracruz': int(13000 * factor_demanda)
    }

    # COSTOS NIVEL 1 (ajustados por XX)
    costos_p_cd_personalizados = {
        k: int(v * factor_costos)
        for k, v in COSTOS_PLANTA_CD.items()
    }

    # COSTOS NIVEL 2 (ajustados por XX)
    costos_cd_c_personalizados = {
        k: int(v * factor_costos)
        for k, v in COSTOS_CD_CIUDAD.items()
    }

    return {
        'plantas': plantas_personalizadas,
        'cds': cds_personalizados,
        'ciudades': ciudades_personalizadas,
        'costos_p_cd': costos_p_cd_personalizados,
        'costos_cd_c': costos_cd_c_personalizados,
        'factores': {
            'demanda': factor_demanda,
            'oferta': factor_oferta,
            'costos': factor_costos
        }
    }

print("‚úÖ Generador multi-nivel creado!")

# PROBAR
escenario_test = generar_escenario_multinivel(61)
print(f"\nüß™ Prueba con XX=61:")
print(f"   Plantas: {sum(escenario_test['plantas'].values()):,}")
print(f"   CDs: {sum(escenario_test['cds'].values()):,}")
print(f"   Demanda: {sum(escenario_test['ciudades'].values()):,}")
```

```python
# ============================================================================
# PROCESADOR MASIVO - TODOS LOS ESTUDIANTES
# ============================================================================
print("\nüéØ PROCESANDO ESCENARIOS PARA TODOS LOS ESTUDIANTES...")
print("=" * 70)

resultados_todos = {}

for i, alumno in enumerate(alumnos_data, 1):
    print(f"\n[{i}/{len(alumnos_data)}] Procesando: {alumno['nombre']}")
    print(f"   üìã Matr√≠cula: {alumno['matricula']} (XX={alumno['XX']})")

    try:
        # Generar escenario
        escenario = escenario_multinivel_estudiante(alumno['matricula'])

        # Resolver
        resultado = optimizar_multinivel(
            escenario['plantas'],
            escenario['cds'],
            escenario['ciudades'],
            escenario['costos_p_cd'],
            escenario['costos_cd_c'],
            10_000_000  # Presupuesto alto para todos
        )

        if resultado:
            resultados_todos[alumno['matricula']] = {
                'alumno': alumno,
                'escenario': escenario,
                'resultado': resultado,
                'status': '‚úÖ √ìptimo'
            }
            print(f"   ‚úÖ Costo: ${resultado['costo_total']:,.0f} | Rutas: {resultado['total_rutas']}")
        else:
            resultados_todos[alumno['matricula']] = {
                'alumno': alumno,
                'status': '‚ùå Sin soluci√≥n'
            }
            print(f"   ‚ùå No se encontr√≥ soluci√≥n √≥ptima")

    except Exception as e:
        print(f"   ‚ö†Ô∏è Error: {str(e)}")
        resultados_todos[alumno['matricula']] = {
            'alumno': alumno,
            'status': f'‚ö†Ô∏è Error: {str(e)}'
        }

print("\n" + "=" * 70)
print(f"üéâ PROCESAMIENTO COMPLETADO!")
print(f"   ‚úÖ Exitosos: {sum(1 for r in resultados_todos.values() if r['status'] == '‚úÖ √ìptimo')}")
print(f"   ‚ùå Fallidos: {sum(1 for r in resultados_todos.values() if '‚ùå' in r['status'])}")
```

```python
# ============================================================================
# üéÆ INTERFAZ INTERACTIVA PARA ESTUDIANTES - MULTI-NIVEL
# ============================================================================
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

print("\nüéÆ CONFIGURANDO INTERFAZ PARA ESTUDIANTES...")
print("=" * 70)

# Crear widgets
dropdown_alumno = widgets.Dropdown(
    options=[f"{a['matricula']} - {a['nombre']}" for a in alumnos_data],
    description='üë§ Alumno:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='600px')
)

slider_presupuesto = widgets.IntSlider(
    value=10_000_000,
    min=5_000_000,
    max=20_000_000,
    step=500_000,
    description='üíµ Presupuesto:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='500px'),
    readout_format=',d'
)

dropdown_escenario = widgets.Dropdown(
    options=['Base', 'Optimista (-10% costos)', 'Pesimista (Tormenta)', 'Estrat√©gico (CD Baj√≠o)'],
    description='üåç Escenario:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='500px')
)

checkbox_cds_ampliados = widgets.Checkbox(
    value=False,
    description='üèóÔ∏è Usar CDs Ampliados (si Infeasible)',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

boton_ejecutar = widgets.Button(
    description='üöÄ EJECUTAR SIMULACI√ìN',
    button_style='success',
    layout=widgets.Layout(width='300px', height='50px'),
    style={'button_color': '#4CAF50', 'font_weight': 'bold'}
)

output_resultados = widgets.Output()

def ejecutar_simulacion(b):
    with output_resultados:
        clear_output(wait=True)

        print("\nüîÑ INICIANDO SIMULACI√ìN...")
        print("=" * 70)

        # Obtener datos del alumno seleccionado
        matricula = dropdown_alumno.value.split(' - ')[0]
        alumno = next(a for a in alumnos_data if a['matricula'] == matricula)

        print(f"üë§ ALUMNO: {alumno['nombre']}")
        print(f"üìã MATR√çCULA: {alumno['matricula']}")
        print(f"üî¢ C√ìDIGO XX: {alumno['XX']}")
        print(f"\n‚öôÔ∏è PAR√ÅMETROS:")
        print(f"   ‚Ä¢ Presupuesto: ${slider_presupuesto.value:,} MXN")
        print(f"   ‚Ä¢ Escenario: {dropdown_escenario.value}")
        print(f"   ‚Ä¢ CDs Ampliados: {'S√≠' if checkbox_cds_ampliados.value else 'No'}")

        # Generar escenario base
        print(f"\nüîß Generando escenario para XX={alumno['XX']}...")
        escenario = generar_escenario_multinivel(alumno['XX'])

        # Aplicar modificaciones seg√∫n tipo de escenario
        if dropdown_escenario.value == 'Optimista (-10% costos)':
            print("   üåü Aplicando escenario OPTIMISTA...")
            escenario['costos_p_cd'] = {k: int(v*0.9) for k,v in escenario['costos_p_cd'].items()}
            escenario['costos_cd_c'] = {k: int(v*0.9) for k,v in escenario['costos_cd_c'].items()}

        elif dropdown_escenario.value == 'Pesimista (Tormenta)':
            print("   ‚õàÔ∏è Aplicando escenario PESIMISTA (Tormenta Sandra)...")
            # Cerrar Veracruz
            demanda_veracruz = escenario['ciudades'].pop('Veracruz', 0)
            escenario['ciudades']['Merida'] = escenario['ciudades'].get('Merida', 0) + int(demanda_veracruz * 0.6)
            escenario['ciudades']['Cancun'] = escenario['ciudades'].get('Cancun', 0) + int(demanda_veracruz * 0.4)
            # Aumentar costos
            escenario['costos_cd_c'] = {k: int(v*1.28) for k,v in escenario['costos_cd_c'].items() if 'Veracruz' not in k}

        elif dropdown_escenario.value == 'Estrat√©gico (CD Baj√≠o)':
            print("   üèóÔ∏è Aplicando escenario ESTRAT√âGICO (Nuevo CD Baj√≠o)...")
            escenario['cds']['CD_Bajio'] = 45000
            # Agregar costos nuevos
            escenario['costos_p_cd'].update({
                ('Planta_CDMX', 'CD_Bajio'): 25,
                ('Planta_Monterrey', 'CD_Bajio'): 50,
                ('Planta_Guadalajara', 'CD_Bajio'): 20,
                ('Planta_Tijuana', 'CD_Bajio'): 70
            })
            escenario['costos_cd_c'].update({
                ('CD_Bajio', c): costo for c, costo in [
                    ('Merida', 180), ('Cancun', 170), ('Toluca', 35),
                    ('Leon', 8), ('SLP', 15), ('Veracruz', 95)
                ] if c in escenario['ciudades']
            })

        # CDs ampliados si est√° activado
        if checkbox_cds_ampliados.value:
            print("   üèóÔ∏è Usando CDs AMPLIADOS...")
            escenario['cds'] = {
                'CD_Norte': 60000,
                'CD_Centro': 80000,
                'CD_Occidente': 50000,
                'CD_Sureste': 40000
            }

        print(f"\nüìä DATOS DEL SISTEMA:")
        print(f"   ‚Ä¢ Plantas: {sum(escenario['plantas'].values()):,} unidades")
        print(f"   ‚Ä¢ CDs: {sum(escenario['cds'].values()):,} capacidad")
        print(f"   ‚Ä¢ Demanda: {sum(escenario['ciudades'].values()):,} unidades")
        print(f"   ‚Ä¢ Balance: {sum(escenario['plantas'].values()) - sum(escenario['ciudades'].values()):+,} unidades")

        # Resolver
        print(f"\nüîß OPTIMIZANDO RED LOG√çSTICA...")
        resultado = optimizar_multinivel(
            escenario['plantas'],
            escenario['cds'],
            escenario['ciudades'],
            escenario['costos_p_cd'],
            escenario['costos_cd_c'],
            slider_presupuesto.value
        )

        if resultado:
            print("\n" + "="*70)
            print("‚úÖ ¬°SOLUCI√ìN √ìPTIMA ENCONTRADA!")
            print("="*70)
            print(f"\nüí∞ COSTO TOTAL: ${resultado['costo_total']:,.0f} MXN")
            print(f"üõ£Ô∏è  RUTAS ACTIVAS: {resultado['total_rutas']}")
            print(f"üì¶ Nivel 1 (Planta‚ÜíCD): {len(resultado['flujos_nivel1'])} rutas")
            print(f"üì¶ Nivel 2 (CD‚ÜíCiudad): {len(resultado['flujos_nivel2'])} rutas")

            print(f"\nüìä DETALLE DE FLUJOS:")
            print(f"\nüè≠ NIVEL 1: Planta ‚Üí Centro Distribuci√≥n")
            print("-" * 70)
            for (p, cd), flujo in sorted(resultado['flujos_nivel1'].items(), key=lambda x: -x[1])[:10]:
                costo_ruta = escenario['costos_p_cd'][(p, cd)] * flujo
                print(f"   ‚Ä¢ {p:20} ‚Üí {cd:15} : {flujo:>8,} unidades (${costo_ruta:>10,.0f})")

            print(f"\nüöö NIVEL 2: Centro Distribuci√≥n ‚Üí Ciudad")
            print("-" * 70)
            for (cd, c), flujo in sorted(resultado['flujos_nivel2'].items(), key=lambda x: -x[1])[:10]:
                costo_ruta = escenario['costos_cd_c'][(cd, c)] * flujo
                print(f"   ‚Ä¢ {cd:15} ‚Üí {c:10} : {flujo:>8,} unidades (${costo_ruta:>10,.0f})")

            print(f"\nüí° AN√ÅLISIS:")
            eficiencia = (sum(escenario['ciudades'].values()) / sum(escenario['plantas'].values())) * 100
            print(f"   ‚Ä¢ Utilizaci√≥n del sistema: {eficiencia:.1f}%")
            print(f"   ‚Ä¢ Costo promedio por unidad: ${resultado['costo_total']/sum(escenario['ciudades'].values()):.2f}")
            print(f"   ‚Ä¢ Presupuesto utilizado: {(resultado['costo_total']/slider_presupuesto.value)*100:.1f}%")

            print(f"\nüìä VISUALIZACI√ìN:")
            print("   Ejecutando visualizaci√≥n interactiva...")
            visualizar_multinivel_mejorado(resultado)

        else:
            print("\n" + "="*70)
            print("‚ùå NO SE ENCONTR√ì SOLUCI√ìN √ìPTIMA")
            print("="*70)
            print("\nüéØ RETO ESPECIAL PARA TI:")
            print("\n   Tu escenario presenta un desaf√≠o real de log√≠stica:")
            print("   La DEMANDA excede la CAPACIDAD disponible.")
            print("\n   üìö Consulta la secci√≥n 'TROUBLESHOOTING' en las")
            print("      instrucciones finales para resolver este problema.")
            print("\n   üí° Tienes 3 opciones estrat√©gicas:")
            print("      1. üèóÔ∏è Ampliar infraestructura (activa el checkbox)")
            print("      2. üìâ Reducir demanda (modifica ciudades)")
            print("      3. üöõ Aumentar plantas (modifica capacidad)")
            print("\n   üèÜ Resolver este problema vale +20 puntos extra!")

boton_ejecutar.on_click(ejecutar_simulacion)

# Mostrar interfaz
print("‚úÖ Interfaz configurada!")
print("\n" + "="*70)
display(HTML("""
<div style='background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            padding: 30px; border-radius: 15px; color: white; text-align: center;
            box-shadow: 0 10px 30px rgba(0,0,0,0.3); margin: 20px 0;'>
    <h2 style='margin: 0; font-size: 28px;'>üöö AMAZON LOGISTICS - EL BUEN FIN 2025</h2>
    <p style='margin: 10px 0 5px 0; font-size: 16px;'>Universidad del SABES - Ingenier√≠a Log√≠stica y Cadena de Valor</p>
    <p style='margin: 5px 0 0 0; font-size: 14px; opacity: 0.9;'>Simulador Interactivo de Optimizaci√≥n Multi-Nivel</p>
    <hr style='border: none; border-top: 2px solid rgba(255,255,255,0.3); margin: 15px 0;'>
    <p style='margin: 0; font-size: 13px;'>Selecciona tu nombre, ajusta par√°metros y ejecuta tu simulaci√≥n personalizada</p>
</div>
"""))

display(dropdown_alumno)
display(slider_presupuesto)
display(dropdown_escenario)
display(checkbox_cds_ampliados)
display(boton_ejecutar)
display(output_resultados)
```
